```python
import torch
import torch.nn as nn
import torch.optim as optim

# TorchRL imports
from torchrl.envs import EnvBase, TransformedEnv, StepCounter, check_env_specs
from torchrl.envs.utils import (
    ExplorationType,
    set_exploration_type,
)
from torchrl.data import (
    CompositeSpec,
    BoundedTensorSpec,
    UnboundedContinuousTensorSpec,
    UnboundedDiscreteTensorSpec,


)
# Import TensorDict and TensorDictBase from the tensordict library
from tensordict import TensorDict, TensorDictBase
from tensordict.nn import TensorDictModule
from tensordict.nn.probabilistic import ProbabilisticTensorDictModule

from torchrl.collectors import SyncDataCollector
from torchrl.objectives import ClipPPOLoss

from torchrl.objectives.value.advantages import GAE

# Graph-specific imports (PyTorch Geometric)
import networkx as nx
import torch_geometric
from torch_geometric.nn import GCNConv, global_mean_pool
# Ensure correct import for Data and Batch
from torch_geometric.data import Data, Batch

# Other standard libraries
import numpy as np
import pandas as pd
import os
import random
import math
import time
import wandb
from scipy.stats import zscore


os.environ['CAPTURE_NONTENSOR_STACK'] = 'False'
print("Set CAPTENSOR_NONTENSOR_STACK environment variable to False.")


# Check if torch_geometric is available
_torch_geometric_available = False
try:
    from torch_geometric.nn import GCNConv, global_mean_pool
    from torch_geometric.data import Data, Batch
    _torch_geometric_available = True
except ImportError:
    print("torch_geometric not found. Please install it to use GNN-based models.")


class FuelpriceenvfeatureGraph():

    def __init__(self, device=torch.device("cpu")):
        """
        Initializes the FuelpriceenvfeatureGraph by loading and preprocessing data.

        Args:
            device (torch.device): The device to place tensors on (default is 'cpu').
        """
        print("Initializing FuelpriceenvfeatureGraph...")
        self.device = device
        self.graph = nx.DiGraph()
        self.graph.graph["graph_attr_1"] = random.random() * 10
        self.graph.graph["graph_attr_2"] = random.random() * 5.

        self.combined_data = None
        self.processed_df = None
        self.date_index = None
        self.feature_columns = None
        self.obs_dim = None
        self.obs_min = None
        self.obs_max = None

        try:
            self._load_data()
            print("\nFuelpriceenvfeatureGraph initialized. Data loaded successfully.")

            # Assertions after _load_data
            assert self.processed_df is not None and not self.processed_df.empty, "processed_df should not be empty after loading and cleaning."
            assert isinstance(self.processed_df.index, pd.DatetimeIndex), "processed_df index should be a DatetimeIndex."
            assert not self.processed_df.isnull().values.any(), "processed_df should not contain any NaN values after cleaning."

            assert self.features_df is not None and not self.features_df.empty, "features_df should not be empty after selection."
            assert len(self.features_df.columns) == len(self.feature_columns), f"features_df should have {len(self.feature_columns)} columns."

            # combined_data_df should have obs_dim (features) + obs_dim (returns) + obs_dim (actions) columns
            expected_combined_cols_df = len(self.feature_columns) * 3 if self.feature_columns is not None else 0
            assert self.combined_data_df is not None, "combined_data_df should not be None."
            # Check shape only if not empty
            if not self.combined_data_df.empty:
                 assert self.combined_data_df.shape[1] == expected_combined_cols_df, f"combined_data_df should have {expected_combined_cols_df} columns, but has {self.combined_data_df.shape[1]}."

            assert self.combined_data is not None, "combined_data tensor should not be None."
            # Check shape only if not empty
            if self.combined_data.size(0) > 0:
                 assert self.combined_data.shape[1] == expected_combined_cols_df, f"combined_data tensor should have {expected_combined_cols_df} columns, but has {self.combined_data.shape[1]}."
                 assert self.combined_data.dtype == torch.float32, "combined_data tensor should have dtype float32."
                 assert self.combined_data.device == self.device, f"combined_data tensor should be on device {self.device}."


            assert self.date_index is not None and not self.date_index.empty, "date_index should not be empty."
            assert len(self.date_index) == self.combined_data.size(0), "date_index length should match combined_data rows."

            assert self.obs_dim is not None and self.obs_dim == len(self.feature_columns), "obs_dim should be set and match feature column count."

            assert self.obs_min is not None and self.obs_max is not None, "obs_min and obs_max should be set."
            assert self.obs_min.shape == (self.obs_dim, 1), f"obs_min should have shape ({self.obs_dim}, 1)."
            assert self.obs_max.shape == (self.obs_dim, 1), f"obs_max should have shape ({self.obs_dim}, 1)."
            assert self.obs_min.dtype == torch.float32 and self.obs_max.dtype == torch.float32, "obs_min and obs_max should have dtype float32."
            assert self.obs_min.device == self.device and self.obs_max.device == self.device, f"obs_min and obs_max should be on device {self.device}."

            print("Assertions after __init__ passed.")

        except Exception as e:
            print(f"Error during FuelpriceenvfeatureGraph initialization or assertions: {e}")
            raise e


    def _load_data(self):
        """
        Loads, preprocesses, and prepares the data for graph representation.
        Handles date indexing, missing values, outliers, calculates returns/actions,
        and creates the combined data tensor.
        """
        import pandas as pd
        import numpy as np
        import time
        # from scipy.stats import zscore # Already imported at the top

        data_path = '/content/drive/MyDrive/deep learning codes/EIAAPI_DOWNLOAD/solutions/mergedata/Cleaneddata.csv'

        # Add a check and a small delay to ensure the file is available
        max_retries = 10
        for i in range(max_retries):
            if os.path.exists(data_path) and os.path.getsize(data_path) > 0:
                print(f"Data file found at {data_path}.")
                break
            else:
                print(f"Attempt {i+1}/{max_retries}: Data file not found or is empty at {data_path}. Waiting...")
                time.sleep(1) # Wait for 1 second before retrying
        else:
            # If loop finishes without finding the file
            print(f"Data file not found at {data_path} after {max_retries} retries. Please ensure the file is created and accessible.")
            raise FileNotFoundError(f"Data file not found at {data_path}")


        print("Loading data...")
        try:
            # Load the data, assuming the first column is the date.
            # Read without setting the index initially.
            dffff = pd.read_csv(data_path, header=0)

            # Assertion: Check if the initial DataFrame is not empty
            assert not dffff.empty, "Initial DataFrame loaded from CSV is empty."

            print(f"Columns in loaded CSV before processing: {dffff.columns.tolist()}")
            print(f"First few rows before date processing:\n{dffff.head()}")

            # Assuming the first column in the CSV is always the date, regardless of its name.
            # Convert the first column to datetime. Handle potential errors.
            date_col_name = dffff.columns[0] # Get the name of the first column
            dffff[date_col_name] = pd.to_datetime(dffff[date_col_name], errors='coerce')

            # Check if the date conversion was successful for at least some rows.
            if dffff[date_col_name].isnull().all():
                raise ValueError("Failed to convert any values in the first column to datetime. Check the date format in the CSV.")

            # Set the converted date column as the index and name it 'Date'.
            dffff = dffff.set_index(date_col_name)
            dffff.index.name = 'Date'
            print("Successfully set the first column as DatetimeIndex and named it 'Date'.")
            print(f"First few rows after setting index:\n{dffff.head()}")

            # Assertion: Check if the index is a DatetimeIndex
            assert isinstance(dffff.index, pd.DatetimeIndex), "DataFrame index should be a DatetimeIndex after conversion."


            # Explicitly filter data to be within the desired date range
            initial_rows = dffff.shape[0]
            dffff = dffff[(dffff.index >= '2002-01-01') & (dffff.index <= '2025-12-31')]
            filtered_rows = dffff.shape[0]

            # Assertion: Check if the DataFrame is not empty after date filtering
            assert not dffff.empty, "DataFrame is empty after date filtering."

            print(f"DataFrame filtered from {initial_rows} rows to {filtered_rows} rows. Date range: {dffff.index.min()} to {dffff.index.max()}")
            print(f"DataFrame shape after date filtering: {dffff.shape}")

            # Handle missing values using forward fill and then backward fill
            dffff = dffff.ffill().bfill()
            print(f"DataFrame shape after ffill and bfill: {dffff.shape}")
            print(f"Number of NaNs after ffill/bfill:\n{dffff.isnull().sum()}")

            # Assertion: Check for remaining NaN values after ffill and bfill
            assert not dffff.isnull().values.any(), "DataFrame should not contain any NaN values after ffill and bfill."


            # Identify numeric columns for cleaning steps (excluding the index)
            numeric_cols = dffff.select_dtypes(include=np.number).columns.tolist()
            print(f"Identified numeric columns for cleaning: {numeric_cols}")


            # Refined Outlier Removal using IQR
            cols_to_skip_outliers = []
            for col in numeric_cols:
                if dffff[col].std(skipna=True) == 0 or dffff[col].isnull().all():
                    print(f"Warning: Column '{col}' has zero standard deviation or is all NaN. Skipping outlier removal for this column.")
                    cols_to_skip_outliers.append(col)

            numeric_cols_for_outliers = [col for col in numeric_cols if col not in cols_to_skip_outliers]

            if numeric_cols_for_outliers:
                print(f"Applying outlier removal to columns: {numeric_cols_for_outliers}")
                Q1 = dffff[numeric_cols_for_outliers].quantile(0.25)
                Q3 = dffff[numeric_cols_for_outliers].quantile(0.75)
                IQR = Q3 - Q1

                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR

                outliers = (dffff[numeric_cols_for_outliers] < lower_bound) | (dffff[numeric_cols_for_outliers] > upper_bound)

                for col in numeric_cols_for_outliers:
                    median_val = dffff.loc[~outliers[col], col].median()
                    if not pd.isna(median_val):
                        dffff.loc[outliers[col], col] = median_val
                    else:
                        print(f"Warning: Could not calculate median for column '{col}' to replace outliers (median is NaN). Skipping outlier replacement for this column.")
                print("Outlier replacement complete.")
            else:
                print("No numeric columns with sufficient variability found for outlier removal.")

            print(f"DataFrame shape after outlier handling: {dffff.shape}")


            # Define the feature columns. These should be the columns *in the data* after setting the date as index.
            self.feature_columns = [
                'USD_SDR', 'OPEC', 'Brent', 'WTI',
                "EER_EPD2DC_PF4_Y05LA_DPG", "EER_EPD2DXL0_PF4_RGC_DPG",
                "EER_EPD2DXL0_PF4_Y35NY_DPG", "EER_EPD2F_PF4_Y35NY_DPG",
                "EER_EPJK_PF4_RGC_DPG", "EER_EPLLPA_PF4_Y44MB_DPG",
                "EER_EPMRR_PF4_Y05LA_DPG", "EER_EPMRU_PF4_RGC_DPG",
                "EER_EPMRU_PF4_Y35NY_DPG"
            ]

            # Select only the specified feature columns and ensure they are numeric
            if not all(col in dffff.columns for col in self.feature_columns):
                 missing = [col for col in self.feature_columns if col not in dffff.columns]
                 raise ValueError(f"Missing feature columns after cleaning: {missing}")

            self.features_df = dffff[self.feature_columns].select_dtypes(include=np.number).copy()
            print(f"Features DataFrame shape: {self.features_df.shape}")
            print(f"Features DataFrame columns: {self.features_df.columns.tolist()}")

            # Assertion: Check that features_df is not empty and has the correct number of columns
            assert not self.features_df.empty, "features_df should not be empty."
            assert len(self.features_df.columns) == len(self.feature_columns), f"features_df should have {len(self.feature_columns)} columns."


            # Calculate returns and actions based on features_df
            if not self.features_df.empty and self.features_df.shape[0] > 1:
                 numberr_np = self.features_df.values

                 def returns(x):
                     x = np.array(x)
                     return np.diff(x, axis=0)

                 RRRR = returns(numberr_np)

                 def actionspace(x):
                     x = np.array(x)
                     differences = np.diff(x, axis=0)
                     yxx = np.zeros_like(differences, dtype=int)
                     yxx[differences > 0] = 2
                     yxx[differences < 0] = 0
                     yxx[differences == 0] = 1
                     return yxx

                 action = actionspace(numberr_np)

                 # Align Indep (returns and actions) with the features_df from the second row onwards
                 indep_df = pd.DataFrame(np.hstack((RRRR, action)), index=self.features_df.index[1:])
                 print(f"Independent variables (returns and actions) DataFrame shape: {indep_df.shape}")

                 # Combine features_df (from the second row onwards) with indep_df
                 self.combined_data_df = pd.concat([self.features_df.iloc[1:], indep_df], axis=1)
                 print(f"Combined data DataFrame shape: {self.combined_data_df.shape}")

                 # Store the date index for the combined data
                 self.date_index = self.combined_data_df.index
                 print(f"Date index for combined data loaded with length: {len(self.date_index)}")

                 # Convert combined_data_df values to torch tensor and move to device
                 self.combined_data = torch.as_tensor(self.combined_data_df.values, dtype=torch.float32, device=self.device)
                 print(f"Combined data tensor shape: {self.combined_data.shape} on device {self.device}")

                 # Assertion: Check the combined_data_df and combined_data tensor
                 expected_combined_cols = len(self.feature_columns) * 3 # Features + Returns + Actions
                 assert not self.combined_data_df.empty, "combined_data_df should not be empty if features_df has > 1 row."
                 assert self.combined_data_df.shape[1] == expected_combined_cols, f"combined_data_df should have {expected_combined_cols} columns."
                 assert self.combined_data.size(0) == self.combined_data_df.shape[0], "combined_data tensor rows should match combined_data_df rows."
                 assert self.combined_data.size(1) == expected_combined_cols, f"combined_data tensor should have {expected_combined_cols} columns."
                 assert self.combined_data.dtype == torch.float32, "combined_data tensor should have dtype float32."
                 assert self.combined_data.device == self.device, f"combined_data tensor should be on device {self.device}."
                 assert len(self.date_index) == self.combined_data.size(0), "date_index length should match combined_data rows."


            else:
                print("Warning: Features DataFrame is empty or has only one row after cleaning and filtering. Cannot calculate returns/actions or create combined data.")
                self.combined_data_df = pd.DataFrame()
                self.date_index = pd.Index([])
                # If features_df is empty or 1 row, combined_data tensor should be empty.
                # The number of columns it *would* have is len(features_df.columns) + len(indep_df.columns) = obs_dim + 2*obs_dim.
                # Let's use 3 * len(self.feature_columns) as the expected column count for an empty tensor's shape[1].
                expected_cols = len(self.feature_columns) * 3 if self.feature_columns is not None else 0
                self.combined_data = torch.empty((0, expected_cols), dtype=torch.float32, device=self.device)
                print(f"Combined data tensor is empty with shape: {self.combined_data.shape}")

                # Assertions for empty case
                assert self.combined_data_df.empty, "combined_data_df should be empty if features_df has <= 1 row."
                assert self.combined_data.size(0) == 0, "combined_data tensor should have 0 rows if features_df has <= 1 row."
                # Check columns only if feature_columns is defined
                if self.feature_columns is not None:
                     assert self.combined_data.size(1) == len(self.feature_columns) * 3, f"Empty combined_data tensor should have {len(self.feature_columns) * 3} columns."
                assert self.combined_data.dtype == torch.float32, "combined_data tensor should have dtype float32."
                assert self.combined_data.device == self.device, f"combined_data tensor should be on device {self.device}."
                assert self.date_index.empty, "date_index should be empty if features_df has <= 1 row."



            # Store the processed DataFrame (the one with Date index and cleaning applied)
            self.processed_df = dffff

            # Calculate observation dimensions and bounds from the features_df
            # obs_dim is the number of features per node
            self.obs_dim = len(self.feature_columns) if self.feature_columns is not None else 0
            if not self.features_df.empty:
                self.obs_min = torch.as_tensor(self.features_df.min().values, dtype=torch.float32, device=self.device).unsqueeze(-1)
                self.obs_max = torch.as_tensor(self.features_df.max().values, dtype=torch.float32, device=self.device).unsqueeze(-1)
                print(f"\nData loading and preprocessing complete.")
                print(f"Observation bounds calculated from features_df: Min = {self.obs_min.cpu().numpy()}, Max = {self.obs_max.cpu().numpy()}")

                # Assertions for obs_dim, obs_min, and obs_max
                assert self.obs_dim == len(self.feature_columns), "obs_dim should match feature column count."
                assert self.obs_min.shape == (self.obs_dim, 1), f"obs_min should have shape ({self.obs_dim}, 1)."
                assert self.obs_max.shape == (self.obs_dim, 1), f"obs_max should have shape ({self.obs_dim}, 1)."
                assert self.obs_min.dtype == torch.float32 and self.obs_max.dtype == torch.float32, "obs_min and obs_max should have dtype float32."
                assert self.obs_min.device == self.device and self.obs_max.device == self.device, f"obs_min and obs_max should be on device {self.device}."

            else:
                print("\nData loading and preprocessing complete, but features_df is empty. Observation bounds are zero.")
                # If obs_dim is 0, min/max should be empty or zero size.
                self.obs_min = torch.zeros((self.obs_dim, 1), dtype=torch.float32, device=self.device)
                self.obs_max = torch.zeros((self.obs_dim, 1), dtype=torch.float32, device=self.device)

                # Assertions for empty features_df case
                assert self.obs_dim == (len(self.feature_columns) if self.feature_columns is not None else 0), "obs_dim should match feature column count (0 if features_df empty)."
                assert self.obs_min.shape == (self.obs_dim, 1), f"obs_min should have shape ({self.obs_dim}, 1)."
                assert self.obs_max.shape == (self.obs_dim, 1), f"obs_max should have shape ({self.obs_dim}, 1)."
                assert self.obs_min.dtype == torch.float32 and self.obs_max.dtype == torch.float32, "obs_min and obs_max should have dtype float32."
                assert self.obs_min.device == self.device and self.obs_max.device == self.device, f"obs_min and obs_max should be on device {self.device}."


        except Exception as e:
            print(f"Error during data loading or preprocessing: {e}")
            raise e

    def get_graph_observation(self):
        """
        Generates a PyTorch Geometric Data object using self.combined_data.
        Uses the first self.obs_dim columns as node features, generates a sequential edge index,
        and calculates graph-level attributes from the returns and actions columns.

        Returns:
            data (Data): The graph data object.
            num_nodes (int): The number of nodes in the graph.
            num_edges (int): The number of edges in the graph.
        """
        # Assertion: Check that self.combined_data is not None and not empty before proceeding
        assert self.combined_data is not None and self.combined_data.size(0) > 0, "combined_data is empty. Cannot generate graph observation."

        # Assertion: Check that self.obs_dim is not None
        assert self.obs_dim is not None, "obs_dim is not set. Call _load_data() first."


        # 1. Extract node features using self.obs_dim
        x = self.combined_data[:, :self.obs_dim].to(self.device) # Ensure x is on the correct device
        num_nodes = x.size(0)

        # Assertion: Check node features tensor shape and device
        assert x.shape == (num_nodes, self.obs_dim), f"Node features tensor x should have shape ({num_nodes}, {self.obs_dim})."
        assert x.device == self.device, f"Node features tensor x should be on device {self.device}."


        # 2. Generate sequential edge index
        if num_nodes > 1:
            src = torch.arange(num_nodes - 1, device=self.device)
            dest = torch.arange(1, num_nodes, device=self.device)
            edge_index = torch.stack([src, dest], dim=0)
            num_edges = edge_index.size(1)
        else:
            edge_index = torch.empty((2, 0), dtype=torch.long, device=self.device)
            num_edges = 0

        # Assertion: Check edge index tensor shape, dtype, and device
        assert edge_index.shape == (2, num_edges), f"Edge index tensor should have shape (2, {num_edges})."
        assert edge_index.dtype == torch.long, "Edge index tensor should have dtype torch.long."
        assert edge_index.device == self.device, f"Edge index tensor should be on device {self.device}."


        # 3. Calculate graph attributes
        returns_start_col = self.obs_dim
        actions_start_col = self.obs_dim * 2
        expected_total_cols_combined = self.obs_dim * 3 # Features + Returns + Actions

        # Check if the combined_data tensor has the expected number of columns for returns and actions
        if self.combined_data.size(1) >= actions_start_col + self.obs_dim:
             mean_returns = torch.mean(self.combined_data[:, returns_start_col:actions_start_col], dim=0).to(self.device)
             mean_actions = torch.mean(self.combined_data[:, actions_start_col:actions_start_col + self.obs_dim], dim=0).to(self.device)
             graph_attributes = torch.cat([mean_returns, mean_actions], dim=0).to(self.device)

        else:
            print(f"Warning: Combined data size ({self.combined_data.size(1)}) is not sufficient to extract returns and actions. Expected at least {actions_start_col + self.obs_dim} columns. Creating zero graph attributes.")
            graph_attributes = torch.zeros(2 * self.obs_dim, device=self.device)

        # Assertion: Check graph attributes tensor shape and device
        # Graph attributes are derived from returns and actions, each of size obs_dim
        assert graph_attributes.shape == (2 * self.obs_dim,), f"Graph attributes tensor should have shape ({2 * self.obs_dim},)."
        assert graph_attributes.device == self.device, f"Graph attributes tensor should be on device {self.device}."


        # 5. Construct the torch_geometric.data.Data object
        data = Data(x=x, edge_index=edge_index, graph_attributes=graph_attributes)

        # Assertion: Check that the returned Data object has the required attributes
        assert hasattr(data, 'x'), "Returned Data object should have 'x' attribute."
        assert hasattr(data, 'edge_index'), "Returned Data object should have 'edge_index' attribute."
        assert hasattr(data, 'graph_attributes'), "Returned Data object should have 'graph_attributes' attribute."

        # Assertion: Check that the number of nodes and edges returned match the shapes
        assert num_nodes == data.x.size(0), "Number of nodes returned should match the size of x."
        assert num_edges == data.edge_index.size(1), "Number of edges returned should match the size of edge_index."


        # 6. Return the Data object, number of nodes, and number of edges
        return data, num_nodes, num_edges


# Test the updated class with assertions
try:
    print("\nTesting FuelpriceenvfeatureGraph initialization and data integrity with assertions:")
    # Set device if CUDA is available, otherwise use CPU
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")
    fuel_env_graph_test_assertions = FuelpriceenvfeatureGraph(device=device) # This will call the updated _load_data with assertions

    print("\nFuelpriceenvfeatureGraph instance created and assertions in __init__ passed.")

    # Test get_graph_observation again, which also contains assertions
    print("\nTesting get_graph_observation with assertions:")
    graph_data, num_nodes, num_edges = fuel_env_graph_test_assertions.get_graph_observation()
    print("\nget_graph_observation test after init:")
    print(f"Graph Data Object: {graph_data}")
    print(f"Number of Nodes: {num_nodes}")
    print(f"Number of Edges: {num_edges}")
    if hasattr(graph_data, 'x'):
        print(f"Node Features shape (x): {graph_data.x.shape}")
    if hasattr(graph_data, 'edge_index'):
        print(f"Edge Index shape: {graph_data.edge_index.shape}")
    if hasattr(graph_data, 'graph_attributes'):
        print(f"Graph Attributes shape: {graph_data.graph_attributes.shape}")

    # Verify device
    print(f"\nOutput tensor devices:")
    if hasattr(graph_data, 'x'):
        print(f"Node features device: {graph_data.x.device}")
    if hasattr(graph_data, 'edge_index'):
        print(f"Edge index device: {graph_data.edge_index.device}")
    if hasattr(graph_data, 'graph_attributes'):
        print(f"Graph attributes device: {graph_data.graph_attributes.device}")

    print("\nAssertions in get_graph_observation passed.")

except Exception as e:
    print(f"Error during FuelpriceenvfeatureGraph test with assertions: {e}")

# Code moved from zF-xOE7JGy1p
# Create an instance of FuelpriceenvfeatureGraph to load and process the data
# The __init__ method now calls _load_data automatically
# Set device if CUDA is available, otherwise use CPU
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
fuel_env_graph = FuelpriceenvfeatureGraph(device=device)

print("\nVerifying FuelpriceenvfeatureGraph instance attributes after initialization:")

# Add assertions in the main execution block
assert fuel_env_graph.processed_df is not None and not fuel_env_graph.processed_df.empty, "Instance attribute processed_df is not populated correctly."
assert isinstance(fuel_env_graph.processed_df.index, pd.DatetimeIndex), "Instance attribute processed_df index is not a DatetimeIndex."

assert fuel_env_graph.combined_data is not None and fuel_env_graph.combined_data.size(0) > 0, "Instance attribute combined_data is not populated correctly."
assert fuel_env_graph.combined_data.dtype == torch.float32, "Instance attribute combined_data dtype is incorrect."
assert fuel_env_graph.combined_data.device == device, f"Instance attribute combined_data device is incorrect (expected {device})."


assert fuel_env_graph.date_index is not None and not fuel_env_graph.date_index.empty, "Instance attribute date_index is not populated correctly."
assert len(fuel_env_graph.date_index) == fuel_env_graph.combined_data.size(0), "Instance attribute date_index length does not match combined_data rows."


assert fuel_env_graph.obs_dim is not None and fuel_env_graph.obs_dim > 0, "Instance attribute obs_dim is not populated correctly."
assert fuel_env_graph.feature_columns is not None and len(fuel_env_graph.feature_columns) == fuel_env_graph.obs_dim, "Instance attribute feature_columns is not populated correctly or does not match obs_dim."


assert fuel_env_graph.obs_min is not None and fuel_env_graph.obs_max is not None, "Instance attributes obs_min or obs_max are not populated correctly."
assert fuel_env_graph.obs_min.shape == (fuel_env_graph.obs_dim, 1), f"Instance attribute obs_min shape is incorrect (expected ({fuel_env_graph.obs_dim}, 1))."
assert fuel_env_graph.obs_max.shape == (fuel_env_graph.obs_dim, 1), f"Instance attribute obs_max shape is incorrect (expected ({fuel_env_graph.obs_dim}, 1))."
assert fuel_env_graph.obs_min.dtype == torch.float32 and fuel_env_graph.obs_max.dtype == torch.float32, "Instance attributes obs_min or obs_max dtype is incorrect."
assert fuel_env_graph.obs_min.device == device and fuel_env_graph.obs_max.device == device, f"Instance attributes obs_min or obs_max device is incorrect (expected {device})."


print("All instance attribute assertions after initialization passed.")


# Optional: Test get_graph_observation again after successful init and assertions
print("\nTesting get_graph_observation after initialization and attribute assertions:")
try:
    graph_data, num_nodes, num_edges = fuel_env_graph.get_graph_observation()
    print("\nget_graph_observation test successful.")
    # Optional prints for verification
    # print(f"Graph Data Object: {graph_data}")
    # print(f"Number of Nodes: {num_nodes}")
    # print(f"Number of Edges: {num_edges}")
    # if hasattr(graph_data, 'x'):
    #     print(f"Node Features shape (x): {graph_data.x.shape}")
    # if hasattr(graph_data, 'edge_index'):
    #     print(f"Edge Index shape: {graph_data.edge_index.shape}")
    # if hasattr(graph_data, 'graph_attributes'):
    #     print(f"Graph Attributes shape: {graph_data.graph_attributes.shape}")
except Exception as e:
    print(f"Error during get_graph_observation test after initialization: {e}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd # Import pandas if it's not already globally available
import matplotlib.cm as cm # Import colormaps

class FuelPricePlots:
    """
    A class to generate plots for different fuel and crude oil prices.
    """
    def __init__(self, dataframe):
        """
        Initializes the FuelPricePlots class with a pandas DataFrame.

        Args:
            dataframe (pd.DataFrame): The DataFrame containing the price data
                                      with a DatetimeIndex.
        """
        if not isinstance(dataframe, pd.DataFrame):
            raise TypeError("Input must be a pandas DataFrame.")
        if not isinstance(dataframe.index, pd.DatetimeIndex):
             print("Warning: DataFrame index is not DatetimeIndex. Attempting conversion.")
             try:
                 dataframe.index = pd.to_datetime(dataframe.index)
             except Exception as e:
                 raise ValueError(f"Could not convert DataFrame index to DatetimeIndex: {e}")

        self.df = dataframe.copy() # Work on a copy to avoid modifying the original

    def plot_crude_oil_prices(self):
        """
        Plots Brent and WTI crude oil prices over time.
        """
        plt.figure(figsize=(14, 7))
        if 'Brent' in self.df.columns:
            plt.plot(self.df.index, self.df['Brent'], label='Brent Crude Oil Price')
        if 'WTI' in self.df.columns:
            plt.plot(self.df.index, self.df['WTI'], label='WTI Crude Oil Price')
        if 'OPEC' in self.df.columns:
             plt.plot(self.df.index, self.df['OPEC'], label='OPEC Crude Oil Price')

        plt.title('Crude Oil Prices Over Time')
        plt.xlabel('Date')
        plt.ylabel('Price')
        plt.legend()
        plt.grid(True)
        plt.show()

    def plot_sdr(self):
        """
        Plots SDR per USD over time.
        """
        plt.figure(figsize=(14, 7))
        if 'USD_SDR' in self.df.columns:
            plt.plot(self.df.index, self.df['USD_SDR'], label='SDR per USD')
        else:
             print("Warning: 'USD_SDR' column not found for plotting.")
             return
        plt.title('SDR per USD Over Time')
        plt.xlabel('Date')
        plt.ylabel('SDR per USD')
        plt.legend()
        plt.grid(True)
        plt.show()

    def plot_fuel_prices(self):
        """
        Plots all daily fuel prices over time on individual sheets with unique colors.
        Assumes columns starting with 'EER_' are fuel prices.
        """
        fuel_cols = [col for col in self.df.columns if col.startswith('EER_')]
        if not fuel_cols:
            print("Warning: No fuel price columns found (columns starting with 'EER_').")
            return

        # Generate a list of unique colors
        colors = cm.get_cmap('tab10', len(fuel_cols)) # Using tab10 colormap for up to 10 distinct colors
        color_list = [colors(i) for i in range(len(fuel_cols))]

        for i, col in enumerate(fuel_cols):
            plt.figure(figsize=(14, 7))
            plt.plot(self.df.index, self.df[col], label=col, color=color_list[i])
            plt.title(f'Daily Fuel Price Over Time - {col}')
            plt.xlabel('Date')
            plt.ylabel('Price')
            plt.legend()
            plt.grid(True)
            plt.show()

# Now, you can run the code in the selected cell (WvHdOC01HVLP) after executing this cell.
```
