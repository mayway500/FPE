```python
#@title Custom Distributions & Flatten modules
import torch
import torch.nn as nn
from torchrl.data import CompositeSpec, DiscreteTensorSpec
from torchrl.modules.distributions import OneHotCategorical # Import OneHotCategorical
from tensordict import TensorDictBase # Import TensorDictBase
from torchrl.data import MultiCategorical # Import MultiCategorical distribution class


# Custom MultiCategorical distribution maker
# This function is designed to be passed to ProbabilisticActor
# It takes the action_spec and policy output (logits) and returns a distribution
# that the ProbabilisticActor can sample from.
def multi_categorical_maker():
    """
    Returns a function that creates a MultiCategorical-like distribution
    from logits and an action specification.
    """
    def make_dist(spec, logits):
        """
        Creates a distribution from logits and a specification.
        Args:
            spec (TensorSpec): The action specification. Expected to be a
                               CompositeSpec with key ('agents', 'action')
                               containing a DiscreteTensorSpec with shape
                               [num_agents, num_individual_actions_features]
                               and a defined nvec based on num_action_categories.
            logits (torch.Tensor): The logits from the policy network.
                                   Expected shape: [batch_size, num_agents, num_individual_actions_features, num_action_categories].

        Returns:
            torch.distributions.Distribution: A distribution object.
        """
        # The logits shape is [batch_size, num_agents, num_individual_actions_features, num_action_categories]
        # OneHotCategorical from torchrl expects logits where the *last* dimension
        # corresponds to the categories for each independent categorical variable.
        # The dimensions before the last one are the event shape.
        # In our case, the event shape is [num_agents, num_individual_actions_features].
        # The logits shape [..., num_agents, num_individual_actions_features, num_action_categories]
        # matches this expectation. The batch dimensions come before the event shape.

        # Create the OneHotCategorical distribution
        # It takes the logits directly.
        try:
             # Pass logits and validate against the spec
             # The spec is the environment's action_spec: CompositeSpec({('agents', 'action'): DiscreteTensorSpec(...)})
             # We need to get the nested spec for the action tensor itself.
             if isinstance(spec, CompositeSpec) and ('agents', 'action') in spec.keys(include_nested=True):
                  action_tensor_spec = spec.get(('agents', 'action'))
                  if not isinstance(action_tensor_spec, DiscreteTensorSpec):
                       print(f"Warning: Nested action spec is not DiscreteTensorSpec, but {type(action_tensor_spec)}.")
                       # Attempt to proceed using the nested spec if it has shape/nvec
                       if not hasattr(action_tensor_spec, 'shape') or not hasattr(action_tensor_spec, 'nvec'):
                            raise TypeError("Nested action spec must be a DiscreteTensorSpec or have shape and nvec attributes.")
                  # else: # Nested spec is DiscreteTensorSpec, proceed

                  # Ensure the logits match the expected categories based on the nested spec's nvec.
                  num_categories_from_logits = logits.shape[-1]
                  if action_tensor_spec.nvec is not None and action_tensor_spec.nvec.numel() > 0:
                       if num_categories_from_logits != action_tensor_spec.nvec.unique().item():
                            print(f"Warning: Logits last dim ({num_categories_from_logits}) does not match nested spec nvec unique value ({action_tensor_spec.nvec.unique().item()}).")
                            # This is a potential issue, but let's proceed using logits shape for categories
                  else:
                       print("Warning: Nested action spec nvec is None or empty. Cannot validate category count from spec.")


                  # The event_shape for OneHotCategorical is inferred from the logits shape, excluding the last dimension.
                  # Logits shape [..., num_agents, num_individual_actions_features, num_action_categories]
                  # Inferred event_shape [..., num_agents, num_individual_actions_features] which should match action_tensor_spec.shape.
                  if logits.shape[-(len(action_tensor_spec.shape) + 1):-1] != action_tensor_spec.shape:
                       print(f"Warning: Logits event shape part ({logits.shape[-(len(action_tensor_spec.shape) + 1):-1]}) does not match nested action spec shape ({action_tensor_spec.shape}).")
                       print("Proceeding, but this may indicate a mismatch in model output shape or spec definition.")


                  dist = OneHotCategorical(logits=logits)
                  return dist

             elif isinstance(spec, DiscreteTensorSpec):
                  # Handle case where action spec is a direct DiscreteTensorSpec (previous attempt)
                  print("Warning: make_dist received a direct DiscreteTensorSpec. Expected a CompositeSpec with key ('agents', 'action').")
                  # Attempt to proceed assuming the DiscreteTensorSpec is the spec for the action tensor
                  action_tensor_spec = spec

                  # Ensure the logits match the expected categories based on the spec's nvec.
                  num_categories_from_logits = logits.shape[-1]
                  if action_tensor_spec.nvec is not None and action_tensor_spec.nvec.numel() > 0:
                       if num_categories_from_logits != action_tensor_spec.nvec.unique().item():
                            print(f"Warning: Logits last dim ({num_categories_from_logits}) does not match spec nvec unique value ({action_tensor_spec.nvec.unique().item()}).")
                            # This is a potential issue, but let's proceed using logits shape for categories
                  else:
                       print("Warning: Action spec nvec is None or empty. Cannot validate category count from spec.")


                  # The event_shape for OneHotCategorical is inferred from the logits shape, excluding the last dimension.
                  # Inferred event_shape [..., num_agents, num_individual_actions_features] which should match action_tensor_spec.shape.
                  if logits.shape[-(len(action_tensor_spec.shape) + 1):-1] != action_tensor_spec.shape:
                       print(f"Warning: Logits event shape part ({logits.shape[-(len(action_tensor_spec.shape) + 1):-1]}) does not match action spec shape ({action_tensor_spec.shape}).")
                       print("Proceeding, but this may indicate a mismatch in model output shape or spec definition.")


                  dist = OneHotCategorical(logits=logits)
                  return dist


             else:
                  raise TypeError(f"make_dist requires a CompositeSpec with key ('agents', 'action') or a direct DiscreteTensorSpec, but got {type(spec)}.")

        except Exception as e:
             print(f"An error occurred within make_dist: {e}")
             # Print detailed traceback for debugging
             import traceback
             traceback.print_exc()
             # Return a dummy distribution or raise the error
             raise # Re-raise the exception


    return make_dist


# Dummy Flatten module (if needed, currently not used based on module definitions)
class Flatten(nn.Module):
    def forward(self, x):
        return x.view(x.size(0), -1)

print("Custom distributions (MultiCategorical-like maker using OneHotCategorical) and Flatten module defined.")
```

```python
#@title Environment
import torch
import torch.nn as nn
import pandas as pd
import os
import networkx as nx
import random
import numpy as np

from tensordict import TensorDict, TensorDictBase
from torchrl.envs import EnvBase
from torchrl.data import Unbounded, Categorical, Composite, DiscreteTensorSpec, MultiCategorical as MultiCategoricalSpec
from torchrl.data.tensor_specs import (
    BoundedTensorSpec,
    UnboundedContinuousTensorSpec,
    CompositeSpec as DataCompositeSpec,
    UnboundedDiscreteTensorSpec,
    DiscreteTensorSpec as DataDiscreteTensorSpec,
    MultiDiscreteTensorSpec as DataMultiDiscreteTensorSpec,
    BoxList as DataBoxList,
    CategoricalBox as DataCategoricalBox,
    UnboundedContinuous as DataUnboundedContinuous
)
from typing import Optional
from torchrl.envs.utils import check_env_specs

# Define the AnFuelpriceEnv class, inheriting from EnvBase
class AnFuelpriceEnv(EnvBase):
    # Accept pre-loaded data_tensor as an argument
    # Add hidden_rnn_dim to the constructor
    def __init__(self, data_tensor: TensorDictBase, num_envs, seed, device, num_agents=13, hidden_rnn_dim=64, **kwargs): # Expect TensorDictBase
        self.episode_length = kwargs.get('episode_length', 10000)
        self.num_agents = num_agents
        self.allow_repeat_data = kwargs.get('allow_repeat_data', False)
        self.num_envs = num_envs
        self.current_data_index = torch.zeros(num_envs, dtype=torch.int64, device=device)
        self.hidden_rnn_dim = hidden_rnn_dim # Store hidden_rnn_dim as an instance attribute


        # Use the pre-loaded data_tensor

        # Ensure combined_data is a TensorDictBase
        if not isinstance(data_tensor, TensorDictBase):
             print("AnFuelpriceEnv.__init__: Error: data_tensor is not a TensorDictBase.")
             raise TypeError("data_tensor must be a TensorDictBase instance.")
        self.combined_data = data_tensor.to(self.device) # Ensure combined_data is on the correct device


        self.device = device

        # Ensure data was loaded successfully before proceeding
        if self.combined_data is None or self.combined_data.batch_size[0] == 0:
             print("AnFuelpriceEnv.__init__: Error: combined_data is None or empty after assignment.") # Debug print
             raise RuntimeError("Data loading failed before environment instantiation or combined_data is empty.")

        # Assuming the first num_agents columns of combined_data are the node features
        # Update node_feature_dim based on the nested 'x' tensor in combined_data if available
        # Access the 'x' data from the combined_data (which is structured with time_steps as batch size)
        x_data_from_combined = self.combined_data.get(('agents', 'data', 'x'))
        if x_data_from_combined is not None and x_data_from_combined.size(0) > 0:
             self.node_feature_dim = x_data_from_combined.shape[-1]
             print(f"AnFuelpriceEnv.__init__: Inferred node_feature_dim from combined_data: {self.node_feature_dim}")
        else:
             self.node_feature_dim = 1 # Default value if nested key not found or data is empty
             print(f"AnFuelpriceEnv.__init__: Warning: Nested ('agents', 'data', 'x') not found in combined_data or combined_data is empty. Using default node_feature_dim: {self.node_feature_dim}")


        self.num_individual_actions = 3 # 3 categories per action feature (down/hold/up)
        self.num_individual_actions_features = 13 # 13 action features per agent

        # Calculate bounds assuming the nested ('agents', 'data', 'x') contains the observation features
        if ('agents', 'data', 'x') in self.combined_data.keys(include_nested=True) and self.combined_data.batch_size[0] > 0:
             # Get the nested tensor, flatten time and env dimensions, compute min/max over features
             full_x_data = self.combined_data.get(('agents', 'data', 'x')) # Shape [num_time_steps, num_agents, node_feature_dim]
             # Flatten time and agent dimensions
             full_x_data_flat = full_x_data.reshape(-1, self.node_feature_dim) # Shape [num_time_steps * num_agents, node_feature_dim]
             self.obs_min = torch.min(full_x_data_flat, dim=0)[0].to(self.device) # Shape [node_feature_dim]
             self.obs_max = torch.max(full_x_data_flat, dim=0)[0].to(self.device) # Shape [node_feature_dim]

             # Unsqueeze to match expected shape in spec [features, 1] if necessary, but spec uses Unbounded for now
             # Let's keep them as [features] for now, as Unbounded spec doesn't enforce last dim 1.
             print(f"AnFuelpriceEnv.__init__: Calculated obs_min/max from full nested ('agents', 'data', 'x'). Shapes: {self.obs_min.shape}, {self.obs_max.shape}")

        else:
             print("AnFuelpriceEnv.__init__: Warning: Nested ('agents', 'data', 'x') not found in combined_data or combined_data is empty. Cannot calculate obs_min/max from data. Using dummy bounds.")
             self.obs_min = torch.zeros(self.node_feature_dim, device=self.device)
             self.obs_max = torch.ones(self.node_feature_dim, device=self.device)


        # Check if bounds were set (will be True if the nested key was found)
        if self.obs_min is None or self.obs_max is None or self.obs_min.numel() == 0 or self.obs_max.numel() == 0:
             print("AnFuelpriceEnv.__init__: Error: Observation bounds (obs_min/obs_max) were not set correctly or are empty after data loading.") # Debug print
             raise RuntimeError("Observation bounds (obs_min/obs_max) were not set correctly after data loading.")

        # Define graph structure (fixed for now, e.g., fully connected or ring)
        self.num_nodes_per_graph = self.num_agents
        # Assuming a fully connected graph for simplicity, adjust if needed.
        # For fully connected, each node connects to every other node.
        # Number of directed edges is num_nodes * (num_nodes - 1)
        if self.num_agents > 1:
             sources = torch.arange(self.num_agents).repeat(self.num_agents - 1)
             targets = torch.arange(self.num_agents).repeat_interleave(self.num_agents - 1)
             mask = sources != targets
             self._fixed_edge_index_single = torch.stack([sources[mask], targets[mask]], dim=0).to(torch.long).to(self.device) # Ensure on device
             self._fixed_num_edges_single = self._fixed_edge_index_single.shape[1]
        else:
             self._fixed_edge_index_single = torch.empty(2, 0, dtype=torch.long, device=self.device) # Ensure on device
             self._fixed_num_edges_single = 0

        print(f"AnFuelpriceEnv.__init__: num_edges_per_graph calculated as {self._fixed_num_edges_single} (assuming fully connected without self-loops).")


        super().__init__(device=device, batch_size=[num_envs])

        self._make_specs()

    # Add the _set_seed method
    def _set_seed(self, seed: Optional[int] = None):
        # Implement seeding logic here if needed
        if seed is not None:
            self.seed = seed
        else:
            self.seed = torch.seed() # Use torch's current seed if none provided
        return seed

    # Implement the actual _is_terminal method
    def _is_terminal(self) -> torch.Tensor:
        """
        Determines if the current state is a terminal state.

        Returns:
            torch.Tensor: A boolean tensor of shape [num_envs] indicating
                          whether each environment is in a terminal state.
        """
        # For now, keeping it as always False, allowing episodes to end only by truncation.
        # Replace this with your specific termination conditions if any.
        terminated = torch.zeros(self.num_envs, dtype=torch.bool, device=self.device)
        return terminated


    # Implement the actual _batch_reward method
    def _batch_reward(self, data_indices: torch.Tensor, tensordict: TensorDictBase) -> TensorDictBase:
        print("AnFuelpriceEnv._batch_reward: Entering _batch_reward method.") # Debug print
        print(f"AnFuelpriceEnv._batch_reward: data_indices: {data_indices}") # Debug print

        original_batch_shape = data_indices.shape # Shape [num_envs]
        # The reward is per agent, shape [num_envs, num_agents, 1]
        expected_reward_shape = torch.Size([original_batch_shape[0], self.num_agents, 1])
        default_reward_tensor = torch.zeros(expected_reward_shape, dtype=torch.float32, device=self.device)

        print(f"AnFuelpriceEnv._batch_reward: default_reward_tensor shape: {default_reward_tensor.shape}") # Debug print


        # Ensure combined_data is available and not empty
        if self.combined_data is None or self.combined_data.batch_size[0] == 0:
            print("AnFuelpriceEnv._batch_reward: Error: combined_data is None or empty. Data loading likely failed.")
            # Return zero reward if data is missing
            # Return a TensorDict with batch_size matching original_batch_shape
            return TensorDict({"rewards": default_reward_tensor}, batch_size=original_batch_shape, device=self.device)


        # Get the reward data from combined_data using the provided data_indices.
        # The data_indices points to the state *after* the transition.
        # The reward for the transition from state t to t+1 is available at index t+1
        # in the combined_data's nested reward_data tensor, assuming the first index
        # of reward_data in combined_data corresponds to the reward for the transition
        # from time step 0 to 1.
        # However, the combined_data was padded at index 0, so the reward for the
        # transition from t to t+1 is at combined_data[t+1].get(('agents', 'reward_data')).
        # The data_indices tensor already contains the index for the *next* state.
        # So, data_indices corresponds to t+1. We can directly use data_indices to
        # get the reward for the transition leading to that state.

        # Ensure data_indices are within the bounds of combined_data
        max_data_index = torch.as_tensor(self.combined_data.batch_size[0] - 1, device=self.device)
        clamped_data_indices = torch.min(data_indices, max_data_index)
        print(f"AnFuelpriceEnv._batch_reward: Clamped data_indices: {clamped_data_indices}") # Debug print")


        # Access the nested reward_data using the clamped data_indices
        # combined_data is a TensorDict with batch_size [num_time_steps]
        # Indexing with clamped_data_indices returns a TensorDict of shape [num_envs]
        # Access the nested ('agents', 'reward_data') key from this result
        # The result should have shape [num_envs, num_agents, 1]
        extracted_reward_data = None # Initialize to None
        try:
             print("AnFuelpriceEnv._batch_reward: Attempting to slice combined_data.") # Debug print
             rewards_td_slice = self.combined_data[clamped_data_indices]
             print(f"AnFuelpriceEnv._batch_reward: rewards_td_slice batch size: {rewards_td_slice.batch_size}") # Debug print
             print(f"AnFuelpriceEnv._batch_reward: rewards_td_slice keys: {rewards_td_slice.keys(include_nested=True)}") # Debug print


             # Add explicit check for the key before getting
             if ('agents', 'reward_data') not in rewards_td_slice.keys(include_nested=True):
                  print(f"AnFuelpriceEnv._batch_reward: Warning: Key ('agents', 'reward_data') not found in sliced tensordict for indices {clamped_data_indices}. Keys found: {rewards_td_slice.keys(include_nested=True)}")
                  # extracted_reward_data remains None, will be handled below
             else:
                 print("AnFuelpriceEnv._batch_reward: Key ('agents', 'reward_data') found in sliced tensordict. Getting tensor.") # Debug print
                 # Use .get(key, default) to safely access the nested tensor and provide a default if it's missing
                 # Correctly access the nested key ('agents', 'reward_data')
                 extracted_reward_data = rewards_td_slice.get(('agents', 'reward_data'), None) # Get the tensor or None


             # Add diagnostic prints here
             print(f"AnFuelpriceEnv._batch_reward: Extracted extracted_reward_data: {extracted_reward_data}")
             if extracted_reward_data is not None:
                  print(f"AnFuelpriceEnv._batch_reward: Type of extracted_reward_data: {type(extracted_reward_data)}")
                  print(f"AnFuelpriceEnv._batch_reward: Shape of extracted_reward_data: {extracted_reward_data.shape}")
                  print(f"AnFuelpriceEnv._batch_reward: Dtype of extracted_reward_data: {extracted_reward_data.dtype}")
                  print(f"AnFuelpriceEnv._batch_reward: Device of extracted_reward_data: {extracted_reward_data.device}")


             # Handle the case where extracted_reward_data is None or has unexpected properties.
             # Convert to expected shape and dtype as a final safeguard.
             # Check if extracted_reward_data is a tensor and if its shape (excluding the first batch dim) matches the expected shape
             if extracted_reward_data is None or not isinstance(extracted_reward_data, torch.Tensor):
                  print(f"AnFuelpriceEnv._batch_reward: Warning: extracted_reward_data is None or not a tensor after extraction. Returning zero reward.")
                  final_rewards_tensor = default_reward_tensor # Use the predefined zero tensor
             elif extracted_reward_data.shape != expected_reward_shape or extracted_reward_data.dtype != torch.float32 or extracted_reward_data.device != self.device:
                 print(f"AnFuelpriceEnv._batch_reward: Warning: Extracted reward data has unexpected properties. Attempting conversion/reshaping.")
                 print(f"  Expected shape: {expected_reward_shape}, Got shape: {extracted_reward_data.shape}")
                 print(f"  Expected dtype: torch.float32, Got dtype: {extracted_reward_data.dtype}")
                 print(f"  Expected device: {self.device}, Got device: {extracted_reward_data.device}")
                 try:
                      # Attempt to convert and reshape
                      # Ensure the shape matches [num_envs, num_agents, 1]
                      final_rewards_tensor = extracted_reward_data.to(self.device).to(torch.float32).reshape(expected_reward_shape)
                      print("AnFuelpriceEnv._batch_reward: Conversion/Reshaping successful.")
                 except Exception as convert_e:
                      print(f"AnFuelpriceEnv._batch_reward: Error during conversion/reshaping: {convert_e}. Returning zero reward.")
                      final_rewards_tensor = default_reward_tensor # Use the predefined zero tensor
             else:
                  # Data is valid, use it directly
                  print(f"AnFuelpriceEnv._batch_reward: Successfully extracted and validated extracted_reward_data shape: {extracted_reward_data.shape}") # Debug print
                  final_rewards_tensor = extracted_reward_data


        except Exception as e:
             print(f"AnFuelpriceEnv._batch_reward: Error during reward extraction from combined_data slice: {e}")
             # Print detailed traceback for debugging the exception
             import traceback
             traceback.print_exc()
             # Return zero reward if extraction fails
             final_rewards_tensor = default_reward_tensor # Use the predefined zero tensor
             print("AnFuelpriceEnv._batch_reward: Returning zero reward due to extraction error.")


        # The `_batch_reward` method should return a TensorDict with batch_size [num_envs]
        # and the reward tensor under the key "rewards" with shape [num_envs, num_agents, 1].
        # Place the extracted/processed reward under the root key 'rewards'
        print(f"AnFuelpriceEnv._batch_reward: Returning TensorDict with 'rewards' key and shape {final_rewards_tensor.shape}") # Debug print
        return TensorDict({"rewards": final_rewards_tensor}, batch_size=original_batch_shape, device=self.device)


    def _step(self, tensordict: TensorDictBase) -> TensorDictBase:
        # print("AnFuelpriceEnv._step: Entering _step method.") # Debug print
        # print(f"AnFuelpriceEnv._step: Input tensordict structure: {tensordict.keys(include_nested=True)}") # Debug print
        # print(f"AnFuelpriceEnv._step: Input tensordict batch size: {tensordict.batch_size}") # Debug print
        # print(f"AnFuelpriceEnv._step: Current data index before increment: {self.current_data_index}") # Debug print")


        # Check if combined_data is available
        if self.combined_data is None or self.combined_data.batch_size[0] == 0:
            print("AnFuelpriceEnv._step: Error: combined_data is None or empty. Data loading likely failed.")
            # Return a tensordict with done flags set to True and dummy values for other keys
            num_envs = tensordict.batch_size[0] if tensordict.batch_size else self.num_envs
            # Ensure dummy edge_index has the correct shape based on _fixed_num_edges_single
            dummy_edge_index = torch.zeros(num_envs, 2, self._fixed_num_edges_single, dtype=torch.int64, device=self.device)

            # Get the observation spec structure to create dummy observation
            dummy_obs_data = {}
            # Iterate through the state_spec structure for the observation part ('agents', 'data')
            if ('agents', 'data') in self.state_spec.keys(include_nested=True):
                 obs_spec_data = self.state_spec[('agents', 'data')]
                 for key, spec in obs_spec_data.items(include_nested=True):
                      # Create dummy tensor with the correct shape, dtype, and device from the spec
                      dummy_shape = torch.Size([num_envs] + list(spec.shape[1:])) # Match batch size, keep rest from spec
                      dummy_dtype = spec.dtype
                      dummy_device = spec.device
                      dummy_obs_data[key] = torch.zeros(dummy_shape, dtype=dummy_dtype, device=dummy_device)


            dummy_obs_td = TensorDict(dummy_obs_data, batch_size=[num_envs], device=self.device)
            # print("AnFuelpriceEnv._step: Created dummy_obs_td:") # Debug print
            # print(dummy_obs_td) # Debug print")


            # Create dummy reward tensor
            # Get shape from reward_spec
            if 'rewards' in self.reward_spec.keys(): # Check for the new key
                 reward_spec_shape = self.reward_spec['rewards'].shape # Use the new key
                 dummy_reward_shape = torch.Size([num_envs] + list(reward_spec_shape[1:]))
                 dummy_reward_dtype = self.reward_spec['rewards'].dtype # Use the new key
                 dummy_reward_device = self.reward_spec['rewards'].device # Use the new key
                 dummy_reward = torch.zeros(dummy_reward_shape, dtype=dummy_reward_dtype, device=dummy_reward_device)
                 # print(f"Created dummy reward tensor with shape {dummy_reward_shape}, dtype {dummy_reward_dtype}, device {dummy_reward_device}") # Debug print
            else:
                 print("AnFuelpriceEnv._step: Warning: 'rewards' not found in reward_spec. Using default dummy reward shape [num_envs, num_agents, 1].")
                 dummy_reward = torch.zeros(num_envs, self.num_agents, 1, dtype=torch.float32, device=self.device)


            # Create dummy recurrent state tensors
            dummy_rnn_state_shape = torch.Size([num_envs, self.num_agents, self.hidden_rnn_dim])
            dummy_rnn_state_dtype = torch.float32
            dummy_rnn_state_device = self.device

            dummy_rnn_hidden_state = torch.zeros(dummy_rnn_state_shape, dtype=dummy_rnn_state_dtype, device=dummy_rnn_state_device)
            dummy_rnn_hidden_state_forecast = torch.zeros(dummy_rnn_state_shape, dtype=dummy_rnn_state_dtype, device=dummy_rnn_state_device)
            dummy_rnn_hidden_state_value = torch.zeros(dummy_rnn_state_shape, dtype=dummy_rnn_state_dtype, device=dummy_rnn_state_device)


            output_tensordict = TensorDict({ # Define output_tensordict here
                "next": Composite({
                     ("agents", "data"): dummy_obs_td,
                     'rewards': dummy_reward, # Use the new key 'rewards' here
                     # Done flags under "next" should match done_spec's shape for "done", "terminated", "truncated"
                     "terminated": torch.ones(num_envs, 1, dtype=torch.bool, device=self.device),
                     "truncated": torch.zeros(num_envs, 1, dtype=torch.bool, device=self.device),
                     "done": torch.ones(num_envs, 1, dtype=torch.bool, device=self.device),
                     # Include placeholders for next recurrent states
                     ('agents', 'rnn_hidden_state'): dummy_rnn_hidden_state,
                     ('agents', 'rnn_hidden_state_forecast'): dummy_rnn_hidden_state_forecast,
                     ('agents', 'rnn_hidden_state_value'): dummy_rnn_hidden_state_value,
                }),
                # Done flags at the root level should match done_spec's shape excluding the last dim
                "terminated": torch.ones(num_envs, dtype=torch.bool, device=self.device),
                "truncated": torch.zeros(num_envs, dtype=torch.bool, device=self.device),
                "done": torch.ones(num_envs, dtype=torch.bool, device=self.device),

                # Include the current observation data in the root level for the collector to use
                # This is what the collector needs to prepare the input for the *next* step's policy/value forward pass.
                # The structure here should match the state_spec's expectation for the current state.
                # Get the current observation data from the input tensordict under ('agents', 'data')
                # Ensure the key exists in the input tensordict before getting
                 ("agents", "data"): tensordict.get(('agents', 'data'), TensorDict({}, batch_size=[num_envs], device=self.device)), # Use empty TD as fallback
                 # Get current RNN states from input tensordict
                 ('agents', 'rnn_hidden_state'): tensordict.get(('agents', 'rnn_hidden_state'), torch.zeros(num_envs, self.num_agents, self.hidden_rnn_dim, dtype=torch.float32, device=self.device)), # Use self.hidden_rnn_dim
                 ('agents', 'rnn_hidden_state_forecast'): tensordict.get(('agents', 'rnn_hidden_state_forecast'), torch.zeros(num_envs, self.num_agents, self.hidden_rnn_dim, dtype=torch.float32, device=self.device)), # Use self.hidden_rnn_dim
                 ('agents', 'rnn_hidden_state_value'): tensordict.get(('agents', 'rnn_hidden_state_value'), torch.zeros(num_envs, self.num_agents, self.hidden_rnn_dim, dtype=torch.float32, device=self.device)), # Use self.hidden_rnn_dim


            }, batch_size=[self.num_envs], device=self.device)

        print("AnFuelpriceEnv._step: Full output_tensordict structure:") # Diagnostic print
        print(output_tensordict) # Diagnostic print


        return output_tensordict

    def _reset(self, tensordict: Optional[TensorDictBase] = None) -> TensorDictBase:
        print("AnFuelpriceEnv._reset: Entering _reset method.") # Debug print
        # Check if combined_data is available
        if self.combined_data is None or self.combined_data.batch_size[0] == 0:
            print("AnFuelpriceEnv._reset: Error: combined_data is None or empty. Data loading failed during environment initialization.")
            raise RuntimeError("Cannot reset environment: Data loading failed during environment initialization.")

        num_envs = self.num_envs # Get num_envs from self for clarity

        # Ensure edge_index_data is always initialized at the beginning of the method
        edge_index_data = torch.empty(num_envs, 2, 0, dtype=torch.int64, device=self.device)
        # print(f"_reset: Initialized edge_index_data shape = {edge_index_data.shape}") # Debug print")


        if self.allow_repeat_data and self.combined_data is not None:
             # Calculate max_start_index based on data length and episode length
             # Max start index allows for at least episode_length steps.
             # If starting at index i, the last step is i + episode_length - 1.
             # The data needed for step i is at index i. The data needed for reward for step i is at index i+1.
             # The state at the end of the episode (after episode_length steps) is at index i + episode_length.
             # Need data up to index i + episode_length.
             # So max start index should be combined_data.batch_size[0] - episode_length - 1 (if reward needed for last step).
             # Let's assume episode_length is the number of *steps* in an episode.
             # An episode of length E has E+1 states (0 to E) and E transitions.
             # If we start at data index `start_idx`, the states are `start_idx` through `start_idx + episode_length`.
             # The rewards are for transitions from `start_idx` to `start_idx + 1`, ..., `start_idx + episode_length - 1` to `start_idx + episode_length`.
             # The reward for step `t` (from state `t` to `t+1`) is at data index `t+1`.
             # So the reward for the last step (from state `start_idx + episode_length - 1` to `start_idx + episode_length`)
             # is at data index `start_idx + episode_length`.
             # The latest data index we need is `start_idx + episode_length`.
             # So, `start_idx + episode_length` must be less than `combined_data.batch_size[0]`.
             # Maximum `start_idx` is `combined_data.batch_size[0] - 1 - episode_length`.

             max_start_index = self.combined_data.batch_size[0] - self.episode_length - 1
             if max_start_index < 0:
                  # If data length is less than episode length + 1, start from index 0 and warn
                  self.current_data_index = torch.zeros(self.num_envs, dtype=torch.int64, device=self.device)
                  print("Warning: Data length is less than episode_length + 1. Starting episodes from index 0.")
             else:
                  # Randomly sample start index for each environment
                  self.current_data_index = torch.randint(0, max_start_index + 1, (self.num_envs,), dtype=torch.int64, device=self.device)

        else:
             # If not allowing repeat data, always start from index 0 for all environments
             self.current_data_index = torch.zeros(self.num_envs, dtype=torch.int64, device=self.device)

        data_indices = self.current_data_index
        # Clamp data indices to prevent out-of-bounds access during the first state extraction
        # Clamp to ensure we don't go beyond the last available data point index
        max_data_index_reset = torch.as_tensor(self.combined_data.batch_size[0] - 1, device=self.device)
        data_indices = torch.min(data_indices, max_data_index_reset)
        print(f"_reset: Data indices for initial state (clamped to {max_data_index_reset}): {data_indices}") # Debug print")


        # Extract the nested state data ('agents', 'data') for the initial time step from combined_data
        # This TensorDict slice should contain 'x', 'edge_index', 'batch', 'time' for the specified indices.
        # The batch_size of the resulting tensordict will be [num_envs].
        if self.combined_data is not None:
             print(f"_reset: Accessing self.combined_data with indices {data_indices} for initial state data.") # Debug print
             # Get the slice of combined_data corresponding to the initial states for each environment
             initial_state_combined_data_slice = self.combined_data[data_indices]
             print(f"_reset: initial_state_combined_data_slice batch size: {initial_state_combined_data_slice.batch_size}") # Debug print
             print(f"_reset: initial_state_combined_data_slice keys: {initial_state_combined_data_slice.keys(include_nested=True)}") # Debug print

             # Extract the nested ('agents', 'data') from this slice
             initial_observation_data_td = initial_state_combined_data_slice.get(('agents', 'data'), None) # Use .get with default

             # Ensure extracted data is a TensorDictBase
             if not isinstance(initial_observation_data_td, TensorDictBase):
                  print("_reset: Warning: Extracted ('agents', 'data') is not a TensorDictBase. Creating dummy data.")
                  # Create dummy data if extraction fails
                  dummy_obs_data = {}
                  if ('agents', 'data') in self.state_spec.keys(include_nested=True):
                       obs_spec_data = self.state_spec[('agents', 'data')]
                       for key, spec in obs_spec_data.items(include_nested=True):
                            # Create dummy tensor based on spec shape and dtype
                            dummy_shape = torch.Size([num_envs] + list(spec.shape[1:]))
                            dummy_dtype = spec.dtype
                            dummy_device = spec.device
                            dummy_obs_data[key] = torch.zeros(dummy_shape, dtype=dummy_dtype, device=dummy_device)
                  initial_observation_data_td = TensorDict(dummy_obs_data, batch_size=[num_envs], device=self.device)
                  print("_reset: Using dummy initial_observation_data_td.")
             else:
                  # Ensure extracted data is on the correct device
                  initial_observation_data_td = initial_observation_data_td.to(self.device)
                  # print(f"_reset: Extracted initial_observation_data_td and moved to device: {initial_observation_data_td.device}") # Debug print
                  # print(f"_reset: Initial observation data keys: {initial_observation_data_td.keys()}") # Debug print
             # Add print statements for the initial_observation_data_td structure
             # print("_reset: Initial observation data TensorDict structure:")
             # print(initial_observation_data_td)


        else:
            print("_reset: Error: combined_data is None when extracting initial state data.") # Debug print
            # Return dummy data for initial state if combined_data is None
            dummy_obs_data = {}
            if ('agents', 'data') in self.state_spec.keys(include_nested=True):
                 obs_spec_data = self.state_spec[('agents', 'data')]
                 for key, spec in obs_spec_data.items(include_nested=True):
                      dummy_shape = torch.Size([num_envs] + list(spec.shape[1:]))
                      dummy_dtype = spec.dtype
                      dummy_device = spec.device
                      dummy_obs_data[key] = torch.zeros(dummy_shape, dtype=dummy_dtype, device=dummy_device)
            initial_observation_data_td = TensorDict(dummy_obs_data, batch_size=[num_envs], device=self.device)
            print("_reset: Using dummy initial_observation_data_td.")


        # Create the output tensordict to return, containing the initial observation and done flags
        # The initial observation should be nested under ("agents", "data")

        output_tensordict = TensorDict({
            # Include the initial observation structure
            ("agents", "data"): initial_observation_data_td,
            # Set initial done, terminated, truncated flags to False at the root level
            "terminated": torch.zeros(self.num_envs, dtype=torch.bool, device=self.device),
            "truncated": torch.zeros(self.num_envs, dtype=torch.bool, device=self.device),
            "done": torch.zeros(self.num_envs, dtype=torch.bool, device=self.device),
            # Add initial recurrent states (zeros) at the root level, matching spec shape [num_envs, num_agents, hidden_rnn_dim]
            # Ensure these keys exist in the state spec before adding
            ('agents', 'rnn_hidden_state'): torch.zeros(self.num_envs, self.num_agents, self.hidden_rnn_dim, dtype=torch.float32, device=self.device), # Use self.hidden_rnn_dim
            ('agents', 'rnn_hidden_state_forecast'): torch.zeros(self.num_envs, self.num_agents, self.hidden_rnn_dim, dtype=torch.float32, device=self.device), # Use self.hidden_rnn_dim
            ('agents', 'rnn_hidden_state_value'): torch.zeros(self.num_envs, self.num_agents, self.hidden_rnn_dim, dtype=torch.float32, device=self.device), # Use self.hidden_rnn_dim
        }, batch_size=[self.num_envs], device=self.device)

        # Add print statement for the final output_tensordict
        print("AnFuelpriceEnv._reset: Full output_tensordict structure:") # Diagnostic print
        print(output_tensordict) # Diagnostic print


        return output_tensordict


    def _make_specs(self):
        print("AnFuelpriceEnv._make_specs: Entering _make_specs.") # Debug print
        # Define the state_spec to match the structure of the tensordict
        # that will be collected by a collector after a step. This tensordict
        # will contain the current state, action, reward, done, and the *next* state.
        # The spec should reflect the structure of the tensordict returned by _reset and _step.
        # The root level contains the current state and done flags.
        # The "next" key contains the next state, reward, and done flags.

        # Define the spec for the nested ('agents', 'data') observation structure
        # This structure should match the TensorDict created in the data preparation cell
        # for the nested ('agents', 'data') key.
        # The batch size of the nested obs_data_spec is the number of environments.
        obs_data_spec = Composite({
             "x": Unbounded( # Node features [num_envs, num_agents, node_feature_dim]
                 shape=torch.Size([self.num_envs, self.num_agents, self.node_feature_dim]), # Use self.node_feature_dim
                 dtype=torch.float32,
                 device=self.device
             ),
             # Use the fixed number of edges for the spec shape
             "edge_index": Unbounded( # Edge indices [num_envs, 2, _fixed_num_edges_single]
                 shape=torch.Size([self.num_envs, 2, self._fixed_num_edges_single]),
                 dtype=torch.int64,
                 device=self.device
             ),
             # Add 'batch' key to the observation spec
             "batch": Unbounded( # Batch tensor [num_envs, num_agents]
                 shape=torch.Size([self.num_envs, self.num_agents]),
                 dtype=torch.int64,
                 device=self.device
             ),
             # Add timestamp to observation spec
             "time": Unbounded( # Timestamp (e.g., data index) [num_envs, 1]
                  shape=torch.Size([self.num_envs, 1]), # Time is per environment
                  dtype=torch.int64, # Using int64 for data index
                  device=self.device
              ),
         },
         # The batch size of the nested obs_data_spec is the number of environments
         batch_size=[self.num_envs],
         device=self.device,
         )
        print(f"AnFuelpriceEnv._make_specs: obs_data_spec defined with shape {obs_data_spec.shape}.") # Debug print


        # Define the state_spec
        self.state_spec = Composite(
             {
                 # Define the keys for the current state, changed 'observation' to 'data'
                 ("agents", "data"): obs_data_spec, # Use the predefined obs_data_spec

                 # Added top-level done, terminated, truncated keys for the current state
                 "done": Categorical(n=2,
                      shape=torch.Size([self.num_envs]), # Shape from _reset and _step output (root level)
                      dtype=torch.bool,
                      device=self.device),
                 "terminated": Categorical(n=2,
                      shape=torch.Size([self.num_envs]), # Shape from _reset and _step output (root level)
                      dtype=torch.bool,
                      device=self.device),
                 "truncated": Categorical(n=2,
                      shape=torch.Size([self.num_envs]), # Shape from _reset and _step output (root level)
                      dtype=torch.bool,
                      device=self.device),

                 # Add recurrent state keys to the state spec
                 # These keys are where the collector will store and retrieve the recurrent states.
                 # The shapes should match the output of the base modules after reshaping
                 # to [num_envs, num_agents, hidden_rnn_dim]. Assuming hidden_rnn_dim=64.
                 ('agents', 'rnn_hidden_state'): Unbounded( # Policy RNN hidden state
                      shape=torch.Size([self.num_envs, self.num_agents, self.hidden_rnn_dim]), # Use self.hidden_rnn_dim
                      dtype=torch.float32,
                      device=self.device
                 ),
                 ('agents', 'rnn_hidden_state_forecast'): Unbounded( # Forecasting RNN hidden state
                      shape=torch.Size([self.num_envs, self.num_agents, self.hidden_rnn_dim]), # Use self.hidden_rnn_dim
                      dtype=torch.float32,
                      device=self.device
                 ),
                 ('agents', 'rnn_hidden_state_value'): Unbounded( # Value RNN hidden state
                      shape=torch.Size([self.num_envs, self.num_agents, self.hidden_rnn_dim]), # Use self.hidden_rnn_dim
                      dtype=torch.float32,
                      device=self.device
                 ),


                 # Define the keys for the next state, nested under "next", changed 'observation' to 'data'
                 "next": Composite({
                      ("agents", "data"): obs_data_spec.clone(), # Use a clone of the predefined obs_data_spec
                      # Also include reward key under the new key 'rewards' under 'next'
                      'rewards': Unbounded(shape=torch.Size([self.num_envs, self.num_agents, 1]), dtype=torch.float32, device=self.device), # Use the new key
                     # Add top-level done, terminated, truncated keys to the "next" composite
                     "done": Categorical(n=2,
                          shape=torch.Size([self.num_envs, 1]), # Corrected shape to match _step output
                          dtype=torch.bool,
                          device=self.device),
                     "terminated": Categorical(n=2,
                          shape=torch.Size([self.num_envs, 1]), # Corrected shape
                           dtype=torch.bool,
                           device=self.device),
                     "truncated": Categorical(n=2,
                          shape=torch.Size([self.num_envs, 1]), # Corrected shape
                      dtype=torch.bool,
                      device=self.device),
                      # Add recurrent state keys to the next state spec
                      # These keys are where the collector will store the next recurrent states.
                      ('agents', 'rnn_hidden_state'): Unbounded(
                           shape=torch.Size([self.num_envs, self.num_agents, self.hidden_rnn_dim]), # Use self.hidden_rnn_dim
                           dtype=torch.float32,
                           device=self.device
                      ),
                      ('agents', 'rnn_hidden_state_forecast'): Unbounded(
                           shape=torch.Size([self.num_envs, self.num_agents, self.hidden_rnn_dim]), # Use self.hidden_rnn_dim
                           dtype=torch.float32,
                           device=self.device
                      ),
                      ('agents', 'rnn_hidden_state_value'): Unbounded(
                           shape=torch.Size([self.num_envs, self.num_agents, self.hidden_rnn_dim]), # Use self.hidden_rnn_dim
                           dtype=torch.float32,
                           device=self.device
                      ),
                 }),
             },
             # The batch size of the state spec is the number of environments
             batch_size=self.batch_size,
             device=self.device,
         )
        print(f"AnFuelpriceEnv._make_specs: State specification defined with single graph per env structure and batch shape {self.state_spec.shape}.") # Debug print


        # Corrected nvec to be a 1D tensor of size num_individual_actions_features with value num_individual_actions
        nvec_list = [self.num_individual_actions] * self.num_individual_actions_features
        nvec_tensor = torch.tensor(nvec_list, dtype=torch.int64, device=self.device)


        # Define the action specification using the MultiCategorical SPEC class from torchrl.data
        # The shape should be [num_agents, num_individual_actions_features] for an unbatched environment
        # nvec should match the shape of the action tensor, excluding batch dimensions.
        # For an unbatched env (batch_size=[]), action shape is [num_agents, num_individual_actions_features].
        # So, nvec should have shape [num_agents, num_individual_actions_features].
        # Repeat the nvec_tensor for each agent.
        nvec_unbatched = nvec_tensor.unsqueeze(0).repeat(self.num_agents, 1) # Shape [num_agents, num_individual_actions_features]


        # Use DiscreteTensorSpec for the action if it's a single discrete value per agent
        # If each agent outputs a single discrete action from a set of categories:
        # self.action_spec_unbatched = Composite(
        #       {('agents','action'): DiscreteTensorSpec(n=self.num_individual_actions, # Total categories for a single action
        #                                               shape=torch.Size([self.num_agents]), # Shape [num_agents] for a single action per agent
        #                                               dtype=torch.int64,
        #                                               device=self.device)},
        #       batch_size=[],
        #       device=self.device
        # )

        # If each agent outputs num_individual_actions_features discrete values:
        # This seems to be the case based on the policy module output structure.
        # Using MultiCategoricalSpec from torchrl.data is appropriate here.
        # The shape of the spec should match the shape of the action tensor excluding batch dimensions.
        # Action tensor shape: [num_agents, num_individual_actions_features]
        self.action_spec_unbatched = Composite(
              {('agents','action'): MultiCategoricalSpec( # Use the MultiCategorical SPEC class
                                                      # The shape here defines the shape of the action tensor for a single environment: [num_agents, num_individual_actions_features]
                                                      shape=torch.Size([self.num_agents, self.num_individual_actions_features]),
                                                      dtype=torch.int64,
                                                      device=self.device,
                                                      nvec=nvec_unbatched # nvec should match the shape of the action tensor
                                                      )},
              batch_size=[], # Unbatched environment has no batch size at the root
              device=self.device
            )


        # print("\nUnbatched Multi-Agent Action specification defined using nested Composites and MultiCategoricalSpec.") # Debug print
        # print(f"Unbatched Environment action_spec: {self.action_spec_unbatched}") # Debug print
        # The batched action_spec is automatically derived by EnvBase
        # print(f"Batched Environment action_spec: {self.action_spec}") # Debug print


        # Define the reward_spec using the new key 'rewards'
        self.reward_spec = Composite(
             {'rewards': Unbounded(shape=torch.Size([self.num_envs, self.num_agents, 1]), dtype=torch.float32, device=self.device)}, # Use the new key
             batch_size=[self.num_envs],
             device=self.device,
        )
        print(f"AnFuelpriceEnv._make_specs: Agent-wise Reward specification defined with batch shape {self.reward_spec.shape}.") # Debug print

        # Define the done_spec to match the structure of the done keys
        # returned by _step.
        self.done_spec = Composite(
            {
                "done": Categorical(n=2,
                     shape=torch.Size([self.num_envs, 1]), # Corrected shape to match _step output
                     dtype=torch.bool,
                     device=self.device),
                "terminated": Categorical(n=2,
                     shape=torch.Size([self.num_envs, 1]), # Corrected shape
                     dtype=torch.bool,
                     device=self.device),
                "truncated": Categorical(n=2,
                     shape=torch.Size([self.num_envs, 1]), # Corrected shape
                      dtype=torch.bool,
                      device=self.device),
            },
            batch_size=[self.num_envs],
            device=self.device,
        )
        print(f"AnFuelpriceEnv._make_specs: Done specification defined with batch shape {self.done_spec.shape}.") # Debug print

        self.state_spec.unlock_(recurse=True)
        self.action_spec.unlock_(recurse=True) # Keep action_spec unlocked as it is batched
        self.reward_spec.unlock_(recurse=True)
        self.done_spec.unlock_(recurse=True)
        print("AnFuelpriceEnv._make_specs: Exiting _make_specs.") # Debug print")


# Instantiate the environment ONLY if data_tensor is available
# Ensure num_envs, seed, device, and num_agents are defined (assuming previous cells ran)
try:
    _ = num_envs
    _ = seed
    _ = device
    _ = num_agents
    _ = data_tensor # Check if data_tensor is defined from the preprocessing step
    _ = hidden_rnn_dim # Check if hidden_rnn_dim is defined for env spec
except NameError:
    print("Please run the hyperparameters cell and the data loading/preprocessing cell first.")
    # Set essential variables to None or default if not defined
    num_envs = 3
    seed = 42
    device = 'cpu'
    num_agents = 13
    data_tensor = None
    hidden_rnn_dim = 64
    print("Setting default values and data_tensor to None due to missing dependencies.")


if data_tensor is not None:
    try:
        # Use the variables defined in the hyperparameters cell and the loaded data_tensor
        # Pass episode_length and other kwargs as needed
        episode_length = locals().get('episode_length', 10) # Use a shorter episode length for testing
        allow_repeat_data = locals().get('allow_repeat_data', True)

        print("\nInstantiating the AnFuelpriceEnv environment...")
        env = AnFuelpriceEnv(data_tensor=data_tensor, num_envs=num_envs, seed=seed, device=device, episode_length=episode_length, num_agents=num_agents, allow_repeat_data=allow_repeat_data, hidden_rnn_dim=hidden_rnn_dim) # Use defined variables and pass hidden_rnn_dim
        print("\nEnvironment instantiated successfully.")

        # Check environment specs
        print("\nChecking environment specs...") # Debug print
        check_env_specs(env) # Commented out for now as it can be noisy
        print("Environment specs checked successfully.") # Debug print

    except Exception as e:
        print(f"\nAn error occurred during environment instantiation or spec check: {e}")
        # Print detailed traceback for debugging
        import traceback
        traceback.print_exc()
        env = None # Set env to None if instantiation fails

else:
    env = None
    print("\nEnvironment not instantiated because data_tensor is None.")


    # @title Refactored Module Definitions and Instantiation
```

```python
import torch
from tensordict import TensorDict
from torchrl.envs import EnvBase
from torchrl.data import BoundedTensorSpec, CompositeSpec, UnboundedContinuousTensorSpec
from torchrl.envs.utils import check_env_specs

class AnFuelpriceEnv(EnvBase):
    def __init__(self, config=None, **kwargs):
        if config is None:
            config = {}
        self._seed = None
        # Initialize base attributes before super().__init__
        self._batch_size = [] # Default batch size
        self._device = kwargs.get("device", torch.device("cpu")) # Default device


        self.num_agents = config.get("num_agents", 2)
        self.max_steps = config.get("max_steps", 100)
        self.current_step = 0
        self.prices = torch.zeros(self.num_agents)
        self.demand_slopes = torch.tensor(config.get("demand_slopes", [-1.0] * self.num_agents))
        self.demand_intercepts = torch.tensor(config.get("demand_intercepts", [10.0] * self.num_agents))
        self.costs = torch.tensor(config.get("costs", [1.0] * self.num_agents))
        self.price_bounds = config.get("price_bounds", (0.0, 10.0))

        # Call super().__init__ after setting attributes it might use
        super().__init__(device=self._device, batch_size=self._batch_size)


    def _make_specs(self):
        # Define observation spec
        observation_spec = CompositeSpec(
            prices=BoundedTensorSpec(
                low=self.price_bounds[0] * torch.ones(self.num_agents),
                high=self.price_bounds[1] * torch.ones(self.num_agents),
                shape=(self.num_agents,),
                dtype=torch.float32,
                device=self.device
            ),
            step=BoundedTensorSpec(
                low=0,
                high=self.max_steps,
                shape=(1,),
                dtype=torch.int64,
                device=self.device
            ),
            shape=()
        )

        # Define action spec - should be nested under 'agents'
        action_spec = CompositeSpec(
            agents=CompositeSpec(
                 action=BoundedTensorSpec(
                     low=self.price_bounds[0] * torch.ones(self.num_agents),
                     high=self.price_bounds[1] * torch.ones(self.num_agents),
                     shape=(self.num_agents,), # Action shape per agent (assuming scalar price per agent)
                     dtype=torch.float32,
                     device=self.device
                 ),
                 shape=() # Inner CompositeSpec for agents
            ),
            shape=() # Outer CompositeSpec for the environment
        )


        # Define reward spec - should be nested under 'agents'
        reward_spec = CompositeSpec(
            agents=CompositeSpec(
                reward_data=UnboundedContinuousTensorSpec(
                    shape=(self.num_agents, 1), # Reward shape per agent
                    dtype=torch.float32,
                    device=self.device
                ),
                shape=()
            ),
            shape=()
        )


        # Define done spec - usually at top level
        done_spec = BoundedTensorSpec(
            low=0,
            high=1,
            shape=(1,),
            dtype=torch.bool,
            device=self.device
        )

        return observation_spec, action_spec, reward_spec, done_spec

    def _reset(self, tensordict):
        self.current_step = 0
        self.prices = torch.rand(self.num_agents) * (self.price_bounds[1] - self.price_bounds[0]) + self.price_bounds[0]

        observation = TensorDict({
            "prices": self.prices.clone(),
            "step": torch.tensor([self.current_step], dtype=torch.int64, device=self.device),
        }, batch_size=[])

        return observation

    def _step(self, tensordict):
        # Actions should be under ('agents', 'action') in the input tensordict
        # Check if the key exists. If not, this is likely during check_env_specs.
        # Simulate a default action in that case.
        actions = tensordict.get(('agents', 'action'), None)

        if actions is None:
            # This is likely during check_env_specs where action is not provided.
            # Simulate a default action (e.g., zeros) for the check.
            print("Warning: ('agents', 'action') not found in input tensordict to _step. Using default action (zeros) for check_env_specs.")
            # Get the expected action shape from the action spec
            action_spec_agents = self.action_spec.get('agents')
            if action_spec_agents is not None and 'action' in action_spec_agents.keys():
                 action_spec_tensor = action_spec_agents.get('action')
                 # Create a zero tensor with the correct shape and device
                 actions = torch.zeros(action_spec_tensor.shape, dtype=action_spec_tensor.dtype, device=self.device)
            else:
                 # If action spec is not correctly defined, fall back to a simple zero tensor
                 print("Error: Could not get action spec to simulate action. Falling back to simple zero tensor.")
                 actions = torch.zeros(self.num_agents, dtype=torch.float32, device=self.device)


        # Ensure actions are on the correct device
        actions = actions.to(self.device)

        # Ensure actions are within bounds if they came from sampling or are simulated
        # Clamp actions to price bounds
        actions = torch.clamp(actions, self.price_bounds[0], self.price_bounds[1])

        self.prices = actions.clone() # Update internal state with new prices
        self.current_step += 1

        # Calculate rewards
        # _batch_reward expects the full tensordict and the actions tensor
        # It returns a TensorDict with {'agents': {'reward_data': reward_tensor}}
        reward_td_output = self._batch_reward(tensordict, actions)

        # Extract reward tensor from the returned TensorDict using the correct nested key
        reward = reward_td_output.get(('agents', 'reward_data'))

        if reward is None:
            print("Warning: 'agents', 'reward_data' key not found in reward_td_output from _batch_reward.")
            print(f"reward_td_output keys: {reward_td_output.keys(True, True)}")
            # Return zero reward for the agents if reward calculation failed
            reward = torch.zeros((self.num_agents, 1), dtype=torch.float32, device=self.device)

        # Determine if done
        done = torch.tensor([self.current_step >= self.max_steps], dtype=torch.bool, device=self.device)

        # Create next observation
        next_observation = TensorDict({
            "prices": self.prices.clone(),
            "step": torch.tensor([self.current_step], dtype=torch.int64, device=self.device),
        }, batch_size=[], device=self.device) # Ensure observation is on device


        # Create output tensordict
        # Include reward and done at the top level of the output tensordict for TorchRL compatibility
        # Include action under the nested ('agents', 'action') key
        output_tensordict = TensorDict({
            "observation": next_observation, # Nested observation TensorDict
            # Reward should be at the top level of the output tensordict with key 'reward'
            # TorchRL's collector expects the reward tensor directly under 'reward'.
            # However, for multi-agent, it expects it nested under ('agents', 'reward').
            # Let's place it under ('agents', 'reward') to match multi-agent conventions.
            ("agents", "reward"): reward, # Nested reward tensor per agent

            "done": done, # Top-level done flag
            ("agents", "action"): actions, # Include action under the nested key in output
             # Also include 'terminated' and 'truncated' flags, which are expected by the collector/GAE
             "terminated": done if self.current_step >= self.max_steps else torch.zeros_like(done), # Assuming termination only happens at max_steps
             "truncated": torch.zeros_like(done), # Assuming no truncation other than reaching max_steps
        }, batch_size=[], device=self.device) # Ensure output tensordict is on device


        return output_tensordict

    def _batch_reward(self, tensordict, actions):
        # Calculate quantities based on prices (actions)
        quantities = self.demand_intercepts + self.demand_slopes.to(self.device) * actions # Ensure demand params are on correct device

        # Ensure quantities are non-negative
        quantities = torch.clamp(quantities, min=0.0)

        # Calculate profits for each agent
        profits = (actions - self.costs.to(self.device)) * quantities # Ensure costs are on correct device

        # Return profits as reward
        # The reward tensor should have shape [num_agents, 1] for a single step's reward
        reward = profits.unsqueeze(-1) # Ensure shape is (num_agents, 1)

        # Wrap the reward tensor in a nested TensorDict structure matching the reward_spec
        # This TensorDict structure should be returned, but its *content* (the reward tensor)
        # will be placed under the ('agents', 'reward') key in the output of _step.
        reward_td_structure = TensorDict({"agents": {"reward_data": reward}}, batch_size=[], device=self.device)


        return reward_td_structure # Return the TensorDict structure containing the reward tensor

    def _set_seed(self, seed):
        rng = torch.manual_seed(seed)
        self._seed = seed
        return rng

# Example usage:
# env = AnFuelpriceEnv()
# check_env_specs(env)
# print("Environment specs are valid!")
```

```python
# @title  Base modules and the combined module_ to accept 'batch' and 'time' as inputs,
import torch.nn as nn
import torch
from torch_geometric.nn import GCNConv
from tensordict.nn import TensorDictModule
import torch.nn.functional as F # Import F for log_softmax
# Removed import of degree function as manual normalization is removed
# from torch_geometric.utils import degree # Import degree function
from torch_geometric.data import Data, Batch # Import Data and Batch for explicit batch handling


# Define the base modules with GCN and RNN
class SimpleMultiAgentPolicyModuleGCN(nn.Module):
     def __init__(self, input_x_dim, num_agents, num_individual_actions_features, num_action_categories, hidden_rnn_dim=64, gcn_hidden_dim=64):
         super().__init__()
         self.num_agents = num_agents
         self.num_individual_actions_features = num_individual_actions_features
         self.num_action_categories = num_action_categories
         self.hidden_rnn_dim = hidden_rnn_dim
         self.gcn_hidden_dim = gcn_hidden_dim # Hidden dimension for GCN

         # GCN layer: maps node features to gcn_hidden_dim
         # Keep internal normalization and self-loops enabled
         self.gcn1 = GCNConv(input_x_dim, gcn_hidden_dim, add_self_loops=True, normalize=True)

         # RNN layer: processes GCN output sequences
         # Input to RNN is [sequence_length, batch_size, input_size]
         # In our case, sequence_length=1 (single step), batch_size = num_envs * num_agents, input_size = gcn_hidden_dim
         self.rnn = nn.GRU(gcn_hidden_dim, hidden_rnn_dim, batch_first=False)

         # Linear layer: maps flattened RNN output to action logits
         # Output is [batch_size, num_agents * num_individual_actions_features * num_action_categories]
         self.linear = nn.Linear(hidden_rnn_dim * num_agents,
                                 num_agents * num_individual_actions_features * num_action_categories)

         # Initial hidden state buffer (internal to the module, not directly used by collector)
         # Shape [num_layers * num_directions, batch_size, hidden_rnn_dim]
         # For a single-layer, unidirectional GRU, this is [1, batch_size, hidden_rnn_dim]
         # Initialize with a dummy batch_size of 1, will be expanded in forward pass
         self.register_buffer("_initial_rnn_hidden_state", torch.zeros(1, 1, self.hidden_rnn_dim))

     # Corrected forward signature to accept inputs as separate arguments including batch and time
     def forward(self, x, edge_index, batch, time, prev_rnn_hidden_state=None):
         # x shape: [num_envs, num_agents, input_x_dim]
         # edge_index shape: [num_envs, 2, num_edges_per_graph]
         # batch shape: [num_envs, num_agents] - This 'batch' is from torchrl/TensorDict, not torch_geometric batch
         # time shape: [num_envs, 1]
         # prev_rnn_hidden_state shape: [num_envs, num_agents, hidden_rnn_dim] (from collector)

         # Ensure inputs are on the correct device
         x = x.to(x.device)
         edge_index = edge_index.to(x.device)
         # torchrl batch maps env_idx -> row index. We need to map node_idx -> env_idx for torch_geometric batch.
         # The 'batch' tensor provided by the environment is already structured as [num_envs, num_agents]
         # where batch[i, j] = i. Flattening this gives [0,0,0,...,1,1,1,..., num_envs-1,...]
         # This is the correct format for torch_geometric.
         batch_tg = batch.reshape(-1).to(x.device) # Use this as torch_geometric batch tensor

         time = time.to(x.device)
         if prev_rnn_hidden_state is not None:
              prev_rnn_hidden_state = prev_rnn_hidden_state.to(x.device)


         num_envs = x.shape[0]
         num_agents = self.num_agents
         batch_size_flat = num_envs * num_agents # Flatten batch size for GCN and RNN input


         # Reshape x for GCN input: [num_envs * num_agents, input_x_dim]
         x_flat = x.reshape(batch_size_flat, -1)

         # Reshape edge_index for GCN input: [num_envs, 2, num_edges_per_graph] -> [2, num_envs * num_edges_per_graph]
         # Adjust edge indices with offset for batched graph
         edge_index_permuted = edge_index.permute(0, 2, 1) # Shape [num_envs, num_edges_per_graph, 2]
         offsets = torch.arange(num_envs, device=x.device).view(-1, 1, 1) * num_agents
         edge_index_offset = edge_index_permuted + offsets
         edge_index_flat = edge_index_offset.reshape(-1, 2).permute(1, 0); # Shape [2, num_envs * num_edges_per_graph]

         # Create a single torch_geometric Batch object
         # Construct a list of Data objects, one per environment
         data_list = []
         # Determine the number of edges per graph dynamically or from a fixed value
         # Assuming edge_index shape is consistent [num_envs, 2, num_edges_per_graph]
         if num_envs > 0 and edge_index.shape[2] > 0:
             num_edges_per_graph = edge_index.shape[2]
             # Iterate through environments and create Data objects
             for i in range(num_envs):
                 # Extract data for env i
                 x_i = x[i].reshape(num_agents, -1) # Shape [num_agents, input_x_dim]
                 edge_index_i = edge_index[i] # Shape [2, num_edges_per_graph]
                 # No need for batch_i here, Batch.from_data_list handles it.
                 data_list.append(Data(x=x_i, edge_index=edge_index_i))
         else:
             # Handle case with no environments or no edges (e.g., initial reset with empty graphs)
             # Create empty Data objects
             for i in range(num_envs):
                 x_i = x[i].reshape(num_agents, -1) # Still need node features even if no edges
                 data_list.append(Data(x=x_i, edge_index=torch.empty(2, 0, dtype=torch.long, device=x.device)))


         # Create Batch object from the list of Data objects
         if len(data_list) > 0:
             batch_tg_obj = Batch.from_data_list(data_list, follow_batch=['edge_index']) # follow_batch is important if using edge_index_batch
             # Extract flattened x, edge_index, and batch from the Batch object
             x_flat_tg = batch_tg_obj.x # Shape [num_envs * num_agents, input_x_dim]
             edge_index_flat_tg = batch_tg_obj.edge_index # Shape [2, num_envs * num_edges_per_graph]
             batch_tg_flat = batch_tg_obj.batch # Shape [num_envs * num_agents]
         else:
             # Handle empty batch case (shouldn't happen with num_envs > 0, but for safety)
             x_flat_tg = torch.empty(0, self.gcn1.in_channels, dtype=x.dtype, device=x.device)
             edge_index_flat_tg = torch.empty(2, 0, dtype=torch.long, device=x.device)
             batch_tg_flat = torch.empty(0, dtype=torch.long, device=x.device)


         # Debugging prints before GCNConv call (can be commented out later)
         # print(f"Policy Module before GCN: x_flat_tg shape={x_flat_tg.shape}, dtype={x_flat_tg.dtype}, device={x_flat_tg.device}, is_contiguous={x_flat_tg.is_contiguous()}")
         # print(f"Policy Module before GCN: edge_index_flat_tg shape={edge_index_flat_tg.shape}, dtype={edge_index_flat_tg.dtype}, device={edge_index_flat_tg.device}, is_contiguous={edge_index_flat_tg.is_contiguous()}")
         # print(f"Policy Module before GCN: batch_tg_flat shape={batch_tg_flat.shape}, dtype={batch_tg_flat.dtype}, device={batch_tg_flat.device}, is_contiguous={batch_tg_flat.is_contiguous()}")


         # Apply GCN layers - Pass the x_flat from Batch, edge_index_flat from Batch, and batch_tg_flat from Batch
         # Input: x_flat_tg [num_envs * num_agents, input_x_dim], edge_index_flat_tg [2, num_envs * num_edges_per_graph], batch_tg_flat [num_envs * num_agents]
         # Allow GCNConv to handle normalization and self-loops internally
         # Pass None for edge_weight explicitly to avoid potential issues
         gcn_output_flat = self.gcn1(x_flat_tg, edge_index_flat_tg, None, batch_tg_flat) # Shape [num_envs * num_agents, gcn_hidden_dim]
         # --- End GCN Call ---


         # Reshape GCN output for RNN: [1, num_envs * num_agents, gcn_hidden_dim]
         # Sequence length = 1 (single step at a time)
         rnn_input = gcn_output_flat.reshape(1, batch_size_flat, self.gcn_hidden_dim)


         # Reshape input state from collector format [num_envs, num_agents, hidden_rnn_dim]
         # to RNN input format [1, num_envs * num_agents, hidden_rnn_dim]
         rnn_input_state = prev_rnn_hidden_state
         if rnn_input_state is not None:
             # Reshape from [num_envs, num_agents, hidden_rnn_dim] to [1, num_envs * num_agents, hidden_rnn_dim]
             rnn_input_state = rnn_input_state.reshape(1, batch_size_flat, self.hidden_rnn_dim)
         else:
             # If prev_rnn_hidden_state is None (first step), initialize with the correct shape
             # using the buffer and expanding to match the current flattened batch size.
             rnn_input_state = self._initial_rnn_hidden_state.expand(1, batch_size_flat, -1).to(x.device)


         # Pass through RNN
         # rnn_output shape: [1, num_envs * num_agents, hidden_rnn_dim]
         # next_rnn_hidden_state_rnn_format shape: [1, num_envs * num_agents, hidden_rnn_dim]
         rnn_output, next_rnn_hidden_state_rnn_format = self.rnn(rnn_input, rnn_input_state.contiguous())


         # Reshape RNN output to [num_envs, num_agents, hidden_rnn_dim] for the linear layer input
         rnn_output_reshaped = rnn_output.reshape(num_envs, num_agents, self.hidden_rnn_dim)

         # Flatten the reshaped RNN output for the linear layer
         flattened_rnn_output = rnn_output_reshaped.reshape(num_envs, -1) # Shape: [num_envs, num_agents * hidden_rnn_dim]


         # Get logits from the linear layer
         # Output shape: [num_envs, num_agents * num_individual_actions_features * num_action_categories]
         flattened_logits = self.linear(flattened_rnn_output)

         # --- Reshape logits to unflattened shape for ProbabilisticActor ---
         # Expected shape: [num_envs, num_agents, num_individual_actions_features, num_action_categories]
         logits = flattened_logits.reshape(
             num_envs,
             self.num_agents,
             self.num_individual_actions_features,
             self.num_action_categories
         )
         # --- End Reshape ---


         # Reshape the next hidden state from RNN format [1, flat_batch, hidden_dim]
         # back to collector format [num_envs, num_agents, hidden_rnn_dim]
         next_rnn_hidden_state = next_rnn_hidden_state_rnn_format.reshape(num_envs, num_agents, self.hidden_rnn_dim)


         # The TensorDictModule expects the output as a tuple matching the out_keys.
         # The policy module outputs action logits (now unflattened) and the next RNN hidden state.
         # The TensorDictModule will handle mapping the unflattened logits to the correct key.
         # The ProbabilisticActor, when given this unflattened tensor at the specified key,
         # should be able to use the distribution maker correctly.

         return logits, next_rnn_hidden_state


# Define the base value module with GCN and RNN
class SimpleMultiAgentValueModuleGCN(nn.Module):
    def __init__(self, input_x_dim, num_agents, hidden_rnn_dim=64, gcn_hidden_dim=64):
        super().__init__()
        self.num_agents = num_agents
        self.hidden_rnn_dim = hidden_rnn_dim
        self.gcn_hidden_dim = gcn_hidden_dim
        self.input_x_dim = input_x_dim # Store input_x_dim

        # Add a linear layer before GCNConv if input_x_dim is not gcn_hidden_dim
        self.pre_gcn_linear = None
        if self.input_x_dim != self.gcn_hidden_dim:
            # print(f"Value Module: Adding pre-GCN linear layer from {self.input_x_dim} to {self.gcn_hidden_dim}") # Debug print
            self.pre_gcn_linear = nn.Linear(self.input_x_dim, self.gcn_hidden_dim)
            gcn_input_dim = self.gcn_hidden_dim
        else:
            gcn_input_dim = self.input_x_dim

        # GCN layer: maps node features to gcn_hidden_dim
        # Keep internal normalization and self-loops enabled
        self.gcn1 = GCNConv(gcn_input_dim, gcn_hidden_dim, add_self_loops=True, normalize=True)

        # RNN layer: processes GCN output sequences
        self.rnn = nn.GRU(gcn_hidden_dim, hidden_rnn_dim, batch_first=False)

        # Linear layer: maps flattened RNN output to a single value estimate per environment
        self.linear = nn.Linear(self.hidden_rnn_dim * self.num_agents, 1)

        # Initial hidden state buffer
        self.register_buffer("_initial_rnn_hidden_state", torch.zeros(1, 1, self.hidden_rnn_dim))


    # Corrected forward signature to accept inputs as separate arguments including batch and time
    def forward(self, x, edge_index, batch, time, prev_rnn_hidden_state=None):
        # x shape: [num_envs, num_agents, input_x_dim]
        # edge_index shape: [num_envs, 2, num_edges_per_graph]
        # batch shape: [num_envs, num_agents] - This 'batch' is from torchrl/TensorDict
        # time shape: [num_envs, 1]
        # prev_rnn_hidden_state shape: [num_envs, num_agents, hidden_rnn_dim] (from collector)

        # Ensure inputs are on the correct device
        x = x.to(x.device)
        edge_index = edge_index.to(x.device)
        batch_tg = batch.reshape(-1).to(x.device) # Use this as torch_geometric batch tensor
        time = time.to(x.device)
        if prev_rnn_hidden_state is not None:
             prev_rnn_hidden_state = prev_rnn_hidden_state.to(x.device)


        num_envs = x.shape[0]
        num_agents = self.num_agents
        batch_size_flat = num_envs * num_agents # Flatten batch size for GCN and RNN input


        # Reshape x: [num_envs * num_agents, input_x_dim]
        x_flat = x.reshape(batch_size_flat, -1)


        # Pass through pre-GCN linear layer if it exists
        if self.pre_gcn_linear is not None:
            # print(f"Value Module: Applying pre-GCN linear layer to x_flat (shape {x_flat.shape})") # Debug print
            x_flat = self.pre_gcn_linear(x_flat)
            # print(f"Value Module: x_flat shape after linear = {x_flat.shape}") # Debug print


        # Reshape edge_index for GCN (with offset)
        edge_index_permuted = edge_index.permute(0, 2, 1)
        # num_edges_per_graph = edge_index_permuted.shape[1] # Not needed here
        offsets = torch.arange(num_envs, device=x.device).view(-1, 1, 1) * num_agents
        edge_index_offset = edge_index_permuted + offsets
        edge_index_flat = edge_index_offset.reshape(-1, 2).permute(1, 0);

        # Create a single torch_geometric Batch object
        # Construct a list of Data objects, one per environment
        data_list = []
        # Determine the number of edges per graph dynamically or from a fixed value
        if num_envs > 0 and edge_index.shape[2] > 0:
             num_edges_per_graph = edge_index.shape[2]
             # Iterate through environments and create Data objects
             for i in range(num_envs):
                 # Extract data for env i
                 x_i = x[i].reshape(num_agents, -1) # Shape [num_agents, input_x_dim]
                 edge_index_i = edge_index[i] # Shape [2, num_edges_per_graph]
                 # No need for batch_i here, Batch.from_data_list handles it.
                 data_list.append(Data(x=x_i, edge_index=edge_index_i))
        else:
            # Handle case with no environments or no edges
            for i in range(num_envs):
                 x_i = x[i].reshape(num_agents, -1) # Still need node features even if no edges
                 data_list.append(Data(x=x_i, edge_index=torch.empty(2, 0, dtype=torch.long, device=x.device)))


        # Create Batch object from the list of Data objects
        if len(data_list) > 0:
             batch_tg_obj = Batch.from_data_list(data_list, follow_batch=['edge_index'])
             # Extract flattened x, edge_index, and batch from the Batch object
             x_flat_tg = batch_tg_obj.x # Shape [num_envs * num_agents, input_x_dim]
             edge_index_flat_tg = batch_tg_obj.edge_index # Shape [2, num_envs * num_edges_per_graph]
             batch_tg_flat = batch_tg_obj.batch # Shape [num_envs * num_agents]
        else:
             # Handle empty batch case
             x_flat_tg = torch.empty(0, self.gcn1.in_channels, dtype=x.dtype, device=x.device)
             edge_index_flat_tg = torch.empty(2, 0, dtype=torch.long, device=x.device)
             batch_tg_flat = torch.empty(0, dtype=torch.long, device=x.device)


        # Debugging prints before GCNConv call (can be commented out later)
        # print(f"Value Module before GCN: x_flat_tg shape={x_flat_tg.shape}, dtype={x_flat_tg.dtype}, device={x_flat_tg.device}, is_contiguous={x_flat_tg.is_contiguous()}")
        # print(f"Value Module before GCN: edge_index_flat_tg shape={edge_index_flat_tg.shape}, dtype={edge_index_flat_tg.dtype}, device={edge_index_flat_tg.device}, is_contiguous={edge_index_flat_tg.is_contiguous()}")
        # print(f"Value Module before GCN: batch_tg_flat shape={batch_tg_flat.shape}, dtype={batch_tg_flat.dtype}, device={batch_tg_flat.device}, is_contiguous={batch_tg_flat.is_contiguous()}")


        # --- Removed Manual GCN Normalization ---
        # Apply GCN layers FIRST - Pass the x_flat_tg, edge_index_flat_tg, and batch_tg_flat
        # Input: x_flat_tg [num_envs * num_agents, gcn_input_dim], edge_index_flat_tg [2, num_envs * num_edges_per_graph], batch_tg_flat [num_envs * num_agents]
        # Allow GCNConv to handle normalization and self-loops internally
        # Pass None for edge_weight explicitly to avoid potential issues
        gcn_output_flat = self.gcn1(x_flat_tg, edge_index_flat_tg, None, batch_tg_flat) # Shape [num_envs * num_agents, gcn_hidden_dim]
        # print(f"Value Module: GCNConv output shape = {gcn_output_flat.shape}") # Debug print


        # Reshape GCN output for RNN: [1, num_envs * num_agents, gcn_hidden_dim]
        # Sequence length = 1 (single step at a time)
        rnn_input = gcn_output_flat.reshape(1, batch_size_flat, self.gcn_hidden_dim);

        # Reshape input state from collector format [num_envs, num_agents, hidden_rnn_dim]
        # to RNN input format [1, num_envs * num_agents, hidden_rnn_dim]
        rnn_input_state = prev_rnn_hidden_state
        if rnn_input_state is not None:
             # Reshape from [num_envs, num_agents, hidden_rnn_dim] to [1, num_envs * num_agents, hidden_rnn_dim]
             rnn_input_state = rnn_input_state.reshape(1, batch_size_flat, self.hidden_rnn_dim);
        else:
             # If prev_rnn_hidden_state is None (first step), initialize with the correct shape
             # using the buffer and expanding to match the current flattened batch size.
             rnn_input_state = self._initial_rnn_hidden_state.expand(1, batch_size_flat, -1).to(x.device);


        # Pass through RNN SECOND
        # rnn_output shape: [1, num_envs * num_agents, hidden_rnn_dim]
        # next_rnn_hidden_state_rnn_format shape: [1, num_envs * num_agents, hidden_rnn_dim]
        rnn_output, next_rnn_hidden_state_rnn_format = self.rnn(rnn_input, rnn_input_state.contiguous());


        # Reshape RNN output to [num_envs, num_agents, hidden_rnn_dim] for the linear layer input
        rnn_output_reshaped = rnn_output.reshape(num_envs, num_agents, self.hidden_rnn_dim);

        # Flatten the reshaped RNN output for the linear layer (for single value per env)
        flattened_rnn_output = rnn_output_reshaped.reshape(num_envs, -1); # Shape: [num_envs, num_agents * hidden_rnn_dim]


        # Get the value from the linear layer
        # Output shape: [num_envs, 1]
        value = self.linear(flattened_rnn_output);


        # Reshape the next hidden state from RNN format [1, flat_batch, hidden_dim]
        # back to collector format [num_envs, num_agents, hidden_rnn_dim]
        next_rnn_hidden_state = next_rnn_hidden_state_rnn_format.reshape(num_envs, num_agents, self.hidden_rnn_dim);


        # The TensorDictModule expects the output as a tuple matching the out_keys.
        # The value module outputs the value estimate and the next RNN hidden state.
        return value, next_rnn_hidden_state


# Define the base forecasting module with GCN and RNN (SimpleMultiAgentForecastingModuleGCN)
class SimpleMultiAgentForecastingModuleGCN(nn.Module):
    """
    A simple GCN-RNN based module for forecasting input features for multiple agents.
    """
    def __init__(self, input_x_dim, num_agents, forecast_horizon, hidden_rnn_dim=64, gcn_hidden_dim=64):
        super().__init__()
        self.num_agents = num_agents
        self.forecast_horizon = forecast_horizon
        self.hidden_rnn_dim = hidden_rnn_dim
        self.input_x_dim = input_x_dim
        self.gcn_hidden_dim = gcn_hidden_dim

        # GCN layer: maps node features to gcn_hidden_dim
        # Keep internal normalization and self-loops enabled
        self.gcn1 = GCNConv(input_x_dim, gcn_hidden_dim, add_self_loops=True, normalize=True)


        # RNN layer: processes GCN output sequences
        self.rnn = nn.GRU(gcn_hidden_dim, hidden_rnn_dim, batch_first=False)

        # Linear layer: maps flattened RNN output to the forecast
        # Output shape: [num_envs, num_agents * input_x_dim * forecast_horizon]
        self.linear = nn.Linear(self.hidden_rnn_dim * self.num_agents,
                                self.num_agents * self.input_x_dim * self.forecast_horizon)

        # Initial hidden state buffer
        self.register_buffer("_initial_rnn_hidden_state", torch.zeros(1, 1, self.hidden_rnn_dim))

    # Corrected forward signature to accept inputs as separate arguments including batch and time
    def forward(self, x, edge_index, batch, time, prev_rnn_hidden_state=None):
        # x shape: [num_envs, num_agents, input_x_dim]
        # edge_index shape: [num_envs, 2, num_edges_per_graph]
        # batch shape: [num_envs, num_agents] - This 'batch' is from torchrl/TensorDict
        # time shape: [num_envs, 1]
        # prev_rnn_hidden_state shape: [num_envs, num_agents, hidden_rnn_dim] (from collector)

        # Ensure inputs are on the correct device
        x = x.to(x.device)
        edge_index = edge_index.to(x.device)
        batch_tg = batch.reshape(-1).to(x.device) # Use this as torch_geometric batch tensor
        time = time.to(x.device)
        if prev_rnn_hidden_state is not None:
             prev_rnn_hidden_state = prev_rnn_hidden_state.to(x.device)

        num_envs = x.shape[0]
        num_agents = self.num_agents
        batch_size_flat = num_envs * num_agents # Flatten batch size for GCN and RNN input


        # Reshape x for GCN: [num_envs * num_agents, input_x_dim]
        x_flat = x.reshape(batch_size_flat, -1);

        # Reshape edge_index for GCN (with offset)
        edge_index_permuted = edge_index.permute(0, 2, 1);
        # num_edges_per_graph = edge_index_permuted.shape[1]; # Not needed here
        offsets = torch.arange(num_envs, device=x.device).view(-1, 1, 1) * num_agents;
        edge_index_offset = edge_index_permuted + offsets;
        edge_index_flat = edge_index_offset.reshape(-1, 2).permute(1, 0);

        # Create a single torch_geometric Batch object
        # Construct a list of Data objects, one per environment
        data_list = []
        # Determine the number of edges per graph dynamically or from a fixed value
        if num_envs > 0 and edge_index.shape[2] > 0:
             num_edges_per_graph = edge_index.shape[2]
             # Iterate through environments and create Data objects
             for i in range(num_envs):
                 # Extract data for env i
                 x_i = x[i].reshape(num_agents, -1) # Shape [num_agents, input_x_dim]
                 edge_index_i = edge_index[i] # Shape [2, num_edges_per_graph]
                 # No need for batch_i here, Batch.from_data_list handles it.
                 data_list.append(Data(x=x_i, edge_index=edge_index_i))
        else:
            # Handle case with no environments or no edges
            for i in range(num_envs):
                 x_i = x[i].reshape(num_agents, -1) # Still need node features even if no edges
                 data_list.append(Data(x=x_i, edge_index=torch.empty(2, 0, dtype=torch.long, device=x.device)))


        # Create Batch object from the list of Data objects
        if len(data_list) > 0:
             batch_tg_obj = Batch.from_data_list(data_list, follow_batch=['edge_index'])
             # Extract flattened x, edge_index, and batch from the Batch object
             x_flat_tg = batch_tg_obj.x # Shape [num_envs * num_agents, input_x_dim]
             edge_index_flat_tg = batch_tg_obj.edge_index # Shape [2, num_envs * num_edges_per_graph]
             batch_tg_flat = batch_tg_obj.batch # Shape [num_envs * num_agents]
        else:
             # Handle empty batch case
             x_flat_tg = torch.empty(0, self.gcn1.in_channels, dtype=x.dtype, device=x.device)
             edge_index_flat_tg = torch.empty(2, 0, dtype=torch.long, device=x.device)
             batch_tg_flat = torch.empty(0, dtype=torch.long, device=x.device)


        # Debugging prints before GCNConv call (can be commented out later)
        # print(f"Forecasting Module before GCN: x_flat_tg shape={x_flat_tg.shape}, dtype={x_flat_tg.dtype}, device={x_flat_tg.device}, is_contiguous={x_flat_tg.is_contiguous()}")
        # print(f"Forecasting Module before GCN: edge_index_flat_tg shape={edge_index_flat_tg.shape}, dtype={edge_index_flat_tg.dtype}, device={edge_index_flat_tg.device}, is_contiguous={edge_index_flat_tg.is_contiguous()}")
        # print(f"Forecasting Module before GCN: batch_tg_flat shape={batch_tg_flat.shape}, dtype={batch_tg_flat.dtype}, device={batch_tg_flat.device}, is_contiguous={batch_tg_flat.is_contiguous()}")


        # --- Removed Manual GCN Normalization ---
        # Apply GCN layers FIRST - Pass the x_flat_tg, edge_index_flat_tg, and batch_tg_flat
        # Input: x_flat_tg [num_envs * num_agents, input_x_dim], edge_index_flat_tg [2, num_envs * num_edges_per_graph], batch_tg_flat [num_envs * num_agents]
        # Allow GCNConv to handle normalization and self-loops internally
        # Pass None for edge_weight explicitly to avoid potential issues
        gcn_output_flat = self.gcn1(x_flat_tg, edge_index_flat_tg, None, batch_tg_flat) # Shape [num_envs * num_agents, gcn_hidden_dim]
        # print(f"Forecasting Module: GCNConv output shape = {gcn_output_flat.shape}") # Debug print


        # Reshape GCN output for RNN: [1, num_envs * num_agents, gcn_hidden_dim]
        # Sequence length = 1 (single step at a time)
        rnn_input = gcn_output_flat.reshape(1, batch_size_flat, self.gcn_hidden_dim);


        # Reshape input state from collector format [num_envs, num_agents, hidden_rnn_dim]
        # to RNN input format [1, num_envs * num_agents, hidden_rnn_dim]
        rnn_input_state = prev_rnn_hidden_state
        if rnn_input_state is not None:
             # Reshape from [num_envs, num_agents, hidden_rnn_dim] to [1, num_envs * num_agents, hidden_rnn_dim]
             rnn_input_state = rnn_input_state.reshape(1, batch_size_flat, self.hidden_rnn_dim);
        else:
             # If prev_rnn_hidden_state is None (first step), initialize with the correct shape
             # using the buffer and expanding to match the current flattened batch size.
             rnn_input_state = self._initial_rnn_hidden_state.expand(1, batch_size_flat, -1).to(x.device);


        # Pass through RNN SECOND
        # rnn_output shape: [1, num_envs * num_agents, hidden_rnn_dim]
        # next_rnn_hidden_state_rnn_format shape: [1, num_envs * num_agents, hidden_rnn_dim]
        rnn_output, next_rnn_hidden_state_rnn_format = self.rnn(rnn_input, rnn_input_state.contiguous());

        # Reshape RNN output to [num_envs, num_agents, hidden_rnn_dim] for the linear layer input
        rnn_output_reshaped = rnn_output.reshape(num_envs, num_agents, self.hidden_rnn_dim);

        # Flatten the reshaped RNN output for the linear layer
        flattened_rnn_output = rnn_output_reshaped.reshape(num_envs, -1);

        # Get the raw forecast output from the linear layer
        # Output shape: [num_envs, num_agents * input_x_dim * forecast_horizon]
        raw_forecast = self.linear(flattened_rnn_output);

        # Reshape the raw forecast output to [num_envs, num_agents, input_x_dim, forecast_horizon]
        forecast = raw_forecast.reshape(num_envs, num_agents, self.input_x_dim, self.forecast_horizon);

        # Reshape the next hidden state from RNN format [1, flat_batch, hidden_dim]
        # back to collector format [num_envs, num_agents, hidden_rnn_dim]
        next_rnn_hidden_state = next_rnn_hidden_state_rnn_format.reshape(num_envs, num_agents, self.hidden_rnn_dim);


        # The TensorDictModule expects the output as a tuple matching the out_keys.
        # The forecasting module outputs the forecast and the next RNN hidden state.
        return forecast, next_rnn_hidden_state

# Rename the base modules to match the names used in the Combined and TensorDictModule classes
PolicyBase = SimpleMultiAgentPolicyModuleGCN
ValueBase = SimpleMultiAgentValueModuleGCN
ForecastingBase = SimpleMultiAgentForecastingModuleGCN

class CombinedPolicyForecastingBase(nn.Module):
    """
    Combines the base policy and forecasting modules.
    Takes state inputs and previous RNN states for both.
    Outputs logits for policy, forecast, and next RNN states for both.
    Updated to accept batch and time inputs and pass them to base modules.
    """
    def __init__(self, policy_module_base: nn.Module, forecasting_module_base: nn.Module, hidden_rnn_dim: int, num_agents: int, device: torch.device):
        super().__init__()
        self.policy_module_base = policy_module_base
        self.forecasting_module_base = forecasting_module_base
        self.hidden_rnn_dim = hidden_rnn_dim
        self.num_agents = num_agents
        self.device = device # Store device

        # Validate input modules are indeed nn.Module instances
        if not isinstance(policy_module_base, nn.Module):
            raise TypeError(f"policy_module_base must be an nn.Module, but got {type(policy_module_base)}")
        if not isinstance(forecasting_module_base, nn.Module):
            raise TypeError(f"forecasting_module_base must be an nn.Module, but got {type(forecasting_module_base)}")


    # Modified forward signature to accept batch and time
    # These argument names must match the string names in the TensorDictModule in_keys tuples
    def forward(self, x: torch.Tensor, edge_index: torch.Tensor, batch: torch.Tensor, time: torch.Tensor, prev_policy_rnn_hidden_state: torch.Tensor = None, prev_forecasting_rnn_hidden_state: torch.Tensor = None) -> tuple[torch.Tensor, torch.Tensor, torch.Tensor, torch.Tensor]:
        """
        Forward pass through the combined policy and forecasting modules.

        Args:
            x (torch.Tensor): Node features. Shape [batch_size, num_nodes, input_x_dim].
            edge_index (torch.Tensor): Edge indices. Shape [batch_size, 2, num_edges_per_graph].
            batch (torch.Tensor): Batch tensor for torch_geometric. Shape [batch_size, num_nodes].
            time (torch.Tensor): Timestamp. Shape [batch_size, 1].
            prev_policy_rnn_hidden_state (torch.Tensor, optional): Previous policy RNN hidden state.
                Shape [batch_size, num_agents, hidden_rnn_dim]. Defaults to None.
            prev_forecasting_rnn_hidden_state (torch.Tensor, optional): Previous forecasting RNN hidden state.
                Shape [batch_size, num_agents, hidden_rnn_dim]. Defaults to None.


        Returns:
            tuple[torch.Tensor, torch.Tensor, torch.Tensor, torch.Tensor]:
                - logits (torch.Tensor): Policy logits. Shape [batch_size, num_agents * num_individual_actions_features * num_action_categories].
                - next_policy_rnn_hidden_state (torch.Tensor): Next policy RNN hidden state. Shape [batch_size, num_agents, hidden_rnn_dim].
                - forecast (torch.Tensor): Forecasting output. Shape [batch_size, num_agents, input_x_dim, forecast_horizon].
                - next_forecasting_rnn_hidden_state (torch.Tensor): Next forecasting RNN hidden state. Shape [batch_size, num_agents, hidden_rnn_dim].
        """
        # Ensure inputs are on the correct device
        x = x.to(self.device)
        edge_index = edge_index.to(self.device)
        batch = batch.to(self.device) # Ensure batch is on the correct device
        time = time.to(self.device) # Ensure time is on the correct device
        if prev_policy_rnn_hidden_state is not None:
            prev_policy_rnn_hidden_state = prev_policy_rnn_hidden_state.to(self.device)
        if prev_forecasting_rnn_hidden_state is not None:
            prev_forecasting_rnn_hidden_state = prev_forecasting_rnn_hidden_state.to(self.device)


        # Pass through policy module, passing arguments by keyword
        # Argument names must match PolicyBase.forward signature
        policy_output, next_policy_rnn_state = self.policy_module_base(
            x=x,
            edge_index=edge_index,
            batch=batch,
            time=time,
            prev_rnn_hidden_state=prev_policy_rnn_hidden_state
        )
        # policy_output shape: [batch_size, num_agents, num_individual_actions_features, num_action_categories] (unflattened)
        # next_policy_rnn_state shape: [batch_size, num_agents, hidden_rnn_dim]

        # Pass through forecasting module, passing arguments by keyword
        # Argument names must match ForecastingBase.forward signature
        forecast, next_forecasting_rnn_state = self.forecasting_module_base(
            x=x,
            edge_index=edge_index,
            batch=batch,
            time=time,
            prev_rnn_hidden_state=prev_forecasting_rnn_hidden_state
        )
        # forecast shape: [batch_size, num_agents, input_x_dim, forecast_horizon]
        # next_forecasting_rnn_state shape: [batch_size, num_agents, hidden_rnn_dim]


        # Return outputs
        # Ensure outputs are on the correct device (should be from the module bases, but double check)
        return policy_output.to(self.device), next_policy_rnn_state.to(self.device), forecast.to(self.device), next_forecasting_rnn_state.to(self.device)

print("Base modules and CombinedPolicyForecastingBase updated to use Batch objects internally and keyword arguments.")
```

```python
#@title Updated Custom Modules (CustomGAE and ManualCollectionModule)

import torch
from tensordict import TensorDict, TensorDictBase
from torchrl.data import CompositeSpec
from torchrl.envs.utils import ExplorationType, set_exploration_type
from torchrl.modules import ProbabilisticActor # Import ProbabilisticActor

# Ensure CombinedPolicyForecastingBase is available
try:
     from __main__ import CombinedPolicyForecastingBase
     print("Imported CombinedPolicyForecastingBase for ManualCollectionModule.")
except ImportError as e:
     print(f"Error importing CombinedPolicyForecastingBase for ManualCollectionModule: {e}. Please ensure the cell defining this class has been executed.")
     CombinedPolicyForecastingBase = None # Set to None to prevent errors if not imported

# Ensure multi_categorical_maker is available
try:
     from __main__ import multi_categorical_maker
     print("Imported multi_categorical_maker for ProbabilisticActor.")
except ImportError as e:
     print(f"Error importing multi_categorical_maker: {e}. Cannot instantiate ProbabilisticActor.")
     multi_categorical_maker = None # Set to None to prevent errors


class ManualCollectionModule(torch.nn.Module):
    """
    A manual collection module to handle the policy forward pass and action sampling
    during data collection, specifically designed for the AnFuelpriceEnv structure.
    This module wraps the combined policy and forecasting module.
    It takes the environment state tensordict as input and outputs the action,
    log probability, value estimate, and next RNN states.
    """
    def __init__(self, combined_module_wrapper: torch.nn.Module, action_spec: CompositeSpec, device: torch.device):
        super().__init__()
        self.combined_module_wrapper = combined_module_wrapper # Wrapped TensorDictModule for policy and forecasting
        self.action_spec = action_spec # Batched action spec from the environment
        self.device = device

        if multi_categorical_maker is not None:
            # Define the ProbabilisticActor to process the 'logits' output from combined_module_wrapper
            # and produce the action and log_prob according to the action_spec.
            # Explicitly pass the in_keys from the wrapped module to ProbabilisticActor.
            self.actor = ProbabilisticActor(
                module=combined_module_wrapper, # Wrap the combined module
                distribution_maker=multi_categorical_maker(), # Use the maker function
                distribution_key=('agents', 'logits'), # Specify the key for logits output by the wrapped module
                return_log_prob=True, # Return log probability
                in_keys=combined_module_wrapper.in_keys, # Explicitly pass the in_keys
                # The action key is inferred from the action_spec, but can be explicit: ('agents', 'action')
                # The log_prob key is inferred, but can be explicit: ('agents', 'action_log_prob')
            ).to(device)
            print("ProbabilisticActor instantiated.")

        else:
             self.actor = None
             print("ProbabilisticActor not instantiated due to missing multi_categorical_maker.")


    def forward(self, tensordict: TensorDictBase) -> TensorDictBase:
        """
        Performs a forward pass through the policy module, samples an action,
        and adds necessary keys to the tensordict for collection.

        Args:
            tensordict (TensorDictBase): The environment state tensordict.
                Expected to contain keys defined by the in_keys of the internal actor module,
                which should match the keys defined in combined_module_wrapper.in_keys.

        Returns:
            TensorDictBase: The input tensordict with added keys, including
                ('agents', 'action'), ('agents', 'action_log_prob'),
                ('agents', 'logits'), 'forecast', and next RNN state keys.
        """
        # Ensure input tensordict is on the correct device
        tensordict = tensordict.to(self.device)

        # Pass the tensordict through the ProbabilisticActor.
        # The ProbabilisticActor will internally call combined_module_wrapper,
        # get the logits and other outputs, sample the action and compute log_prob,
        # and add the action and log_prob (and logits, forecast, next RNN states)
        # to the tensordict.
        if self.actor is not None:
             output_tensordict = self.actor(tensordict)
             output_tensordict = output_tensordict.to(self.device)
             return output_tensordict
        else:
             print("ManualCollectionModule.forward: Actor is None. Skipping forward pass and returning input tensordict.")
             return tensordict


print("ManualCollectionModule class defined.")


#@title Updated Custom GAE

import torch
from tensordict import TensorDictBase # Import TensorDictBase from tensordict directly

class CustomGAE(torch.nn.Module):
    """
    Custom GAE implementation tailored for the AnFuelpriceEnv data structure.
    Calculates advantages and value targets based on collected trajectories,
    handling nested keys for reward, done flags, and value estimates.

    This version is updated to handle potentially flattened keys
    like ('agents', 'reward') directly if the data format changes,
    although the logic here still primarily expects the nested structure
    from the collector output.
    """
    def __init__(self, gamma: float, lmbda: float): # Removed value_network argument
        super().__init__()
        self.gamma = gamma
        self.lmbda = lmbda

    def forward(self, tensordict: TensorDictBase) -> TensorDictBase:
        """
        Computes advantages and value targets for a batch of trajectories.

        Args:
            tensordict (TensorDictBase): The collected tensordict containing
                trajectories with keys for reward, done flags, and value estimates.
                Expected keys:
                - ('next', ('agents', 'reward')) OR ('agents', 'reward') at the end of trajectory: Reward at each step.
                - 'terminated': Termination flag at each step.
                - 'truncated': Truncation flag at each step.
                - ('agents', 'state_value'): Value estimate of the state *at the start of the step*.
                - ('next', ('agents', 'state_value')) OR ('agents', 'state_value') at the end of the trajectory: Value estimate of the state *at the end of the step*.
                  This should include the value estimate for the state after the last step.

        Returns:
            TensorDictBase: The input tensordict with 'advantage' and 'value_target' keys added.
        """
        # Ensure data is on the correct device (matching the GAE module's device)
        # This is important if the GAE module is moved to GPU but data is on CPU.
        # The input tensordict should already be on the correct device from the training loop,
        # but adding .to(self.device) here again is harmless and adds robustness.
        tensordict = tensordict.to(self.device)

        # Extract necessary tensors and ensure they are on the correct device
        # Try both nested and flattened keys for robustness, prioritizing nested from collector
        rewards = tensordict.get(('next', ('agents', 'reward')), tensordict.get(('agents', 'reward'), None))
        if rewards is not None: rewards = rewards.to(self.device)

        terminated = tensordict.get('terminated')
        if terminated is not None: terminated = terminated.to(self.device)

        truncated = tensordict.get('truncated')
        if truncated is not None: truncated = truncated.to(self.device)

        values = tensordict.get(('agents', 'state_value'))
        if values is not None: values = values.to(self.device)

        # Special handling for next_values: need the value of the state *after* the last step in the trajectory.
        # This is expected at ('next', ('agents', 'state_value')) at the last index,
        # or potentially at the root ('agents', 'state_value') if the tensordict is not sliced
        # and represents a single step + next state.
        # Based on collector output structure, it should be at ('next', ('agents', 'state_value'))
        # within the collected batch tensordict.
        next_values_full_trajectory = tensordict.get(('next', ('agents', 'state_value')), None)
        if next_values_full_trajectory is not None: next_values_full_trajectory = next_values_full_trajectory.to(self.device)


        # Check if required keys were found
        if rewards is None:
             raise KeyError("CustomGAE: Could not find rewards at ('next', ('agents', 'reward')) or ('agents', 'reward').")
        if terminated is None:
             raise KeyError("CustomGAE: Could not find 'terminated' flag.")
        if truncated is None:
             raise KeyError("CustomGAE: Could not find 'truncated' flag.")
        if values is None:
             raise KeyError("CustomGAE: Could not find values at ('agents', 'state_value').")
        if next_values_full_trajectory is None:
             raise KeyError("CustomGAE: Could not find next_values at ('next', ('agents', 'state_value')).")


        # Combine terminated and truncated flags into a single done flag for GAE calculation
        # The done flags are for the state at time t+1, corresponding to the transition from time t.
        # In the collected data (shape [num_envs, frames_per_batch]), the flags at index `t`
        # are for the state at time `t+1`.
        dones = terminated | truncated # Shape [num_envs, frames_per_batch]
        # dones is already on the correct device because terminated and truncated were moved


        # GAE calculation: Iterate backwards from T-1 to 0.
        # advantage_t = reward_{t+1} + gamma * (1 - done_{t+1}) * value_{t+1} - value_t

        # We need rewards_{1...T}, done_{1...T}, values_{0...T-1}, next_values_{1...T}.
        # From collected data (shape [num_envs, frames_per_batch]), we have:
        # rewards_{1...T} are at index 0...frames_per_batch-1 in `rewards`. Shape [num_envs, frames_per_batch, num_agents, 1]
        # done_{1...T} are at index 0...frames_per_batch-1 in `dones`. Shape [num_envs, frames_per_batch]
        # values_{0...T-1} are at index 0...frames_per_batch-1 in `values`. Shape [num_envs, frames_per_batch, num_agents, 1]
        # next_values_{1...T} need careful indexing.
        # For t = 0 to T-2, next_value_{t+1} is `values[:, t+1]`.
        # For t = T-1, next_value_T is `next_values_full_trajectory[:, -1]`.

        # Create a tensor for next_values that aligns with the rewards, dones, and values tensors.
        # It should have shape [num_envs, frames_per_batch, num_agents, 1].
        # For indices 0 to frames_per_batch-2, use values[:, 1:].
        # For the last index (frames_per_batch-1), use next_values_full_trajectory[:, -1:].

        # Need to handle the batch shape potentially being more than 2 dimensions (e.g., if num_envs > 1)
        # The time dimension is assumed to be the second last dimension in the collected batch.
        # Let batch_dims = tensordict.batch_dims # Number of batch dimensions, typically 2 ([num_envs, frames_per_batch])
        # time_dim_idx = batch_dims - 1 # Index of the time dimension

        # Let's assume batch shape is [num_envs, frames_per_batch] for now.
        # Indexing for the time dimension is the second dimension (index 1).

        # Get values for time steps 1 to T-1
        values_t_plus_1_slice = values[:, 1:] # Shape [num_envs, frames_per_batch-1, num_agents, 1]

        # Get the last next value (for time step T)
        next_value_T = next_values_full_trajectory[:, -1:] # Shape [num_envs, 1, num_agents, 1]

        # Concatenate to get all next values from time step 1 to T
        next_values_aligned = torch.cat([values_t_plus_1_slice, next_value_T], dim=1)
        # Shape [num_envs, frames_per_batch, num_agents, 1] - This is values_{1...T}


        # Ensure done flags are broadcastable to agent dimension
        dones_expanded = dones.unsqueeze(-1).unsqueeze(-1).expand_as(rewards) # Shape [num_envs, frames_per_batch, num_agents, 1]


        # Calculate deltas: delta_t = reward_{t+1} + gamma * (1 - done_{t+1}) * value_{t+1} - value_t
        # Using our aligned tensors:
        # rewards: rewards_{1...T}
        # dones_expanded: done_{1...T} (expanded)
        # next_values_aligned: value_{1...T} (extracted from values and next_values_full_trajectory)
        # values: value_{0...T-1}

        # The delta calculation needs:
        # rewards[:, t] (reward_{t+1})
        # dones_expanded[:, t] (done_{t+1} expanded)
        # next_values_aligned[:, t] (value_{t+1})
        # values[:, t] (value_t)

        deltas = rewards + self.gamma * (1 - dones_expanded.float()) * next_values_aligned - values
        # Shape of deltas: [num_envs, frames_per_batch, num_agents, 1]


        # Calculate advantages using discounted sum of deltas (backward pass)
        advantages = torch.zeros_like(deltas) # Shape [num_envs, frames_per_batch, num_agents, 1]
        last_advantage = torch.zeros_like(advantages[:, 0]) # Shape [num_envs, num_agents, 1] # Represents advantage_{t+1}

        # Iterate backwards through time steps from T-1 down to 0
        for t in reversed(range(rewards.shape[1])):
             # Current delta is deltas[:, t]
             # The done flag at index t is for the state at time t+1.
             # This flag determines if advantage_{t+1} (last_advantage) should be zeroed out.
             # done_t_plus_1_expanded = dones_expanded[:, t] # Shape [num_envs, num_agents, 1]

             # GAE formula: advantages_t = delta_t + gamma * lmbda * (1 - done_{t+1}) * advantages_{t+1}
             # Using our indexing: advantages[:, t] = deltas[:, t] + gamma * self.lmbda * (1 - dones_expanded[:, t].float()) * last_advantage # Corrected lmbda access

             current_advantage = deltas[:, t] + self.gamma * self.lmbda * (1 - dones_expanded[:, t].float()) * last_advantage
             advantages[:, t] = current_advantage
             last_advantage = current_advantage # Update last_advantage for the next iteration


        # Value targets: value_target_t = advantage_t + value_t
        value_targets = advantages + values

        # Add advantages and value targets to the input tensordict
        tensordict.set('advantage', advantages)
        tensordict.set('value_target', value_targets)

        return tensordict

print("CustomGAE class defined.")
```

```python
#@title CustomeGAE
import torch
from tensordict import TensorDictBase # Import TensorDictBase from tensordict directly

class CustomGAE(torch.nn.Module):
    """
    Custom GAE implementation tailored for the AnFuelpriceEnv data structure.
    Calculates advantages and value targets based on collected trajectories,
    handling nested keys for reward, done flags, and value estimates.

    This version is updated to handle potentially flattened keys
    like ('agents', 'reward') directly if the data format changes,
    although the logic here still primarily expects the nested structure
    from the collector output.
    """
    def __init__(self, gamma: float, lmbda: float):
        super().__init__()
        self.gamma = gamma
        self.lmbda = lmbda

    def forward(self, tensordict: TensorDictBase) -> TensorDictBase:
        """
        Computes advantages and value targets for a batch of trajectories.

        Args:
            tensordict (TensorDictBase): The collected tensordict containing
                trajectories with keys for reward, done flags, and value estimates.
                Expected keys:
                - ('next', ('agents', 'reward')) OR ('agents', 'reward') at the end of trajectory: Reward at each step.
                - 'terminated': Termination flag at each step.
                - 'truncated': Truncation flag at each step.
                - ('agents', 'state_value'): Value estimate of the state *at the start of the step*.
                - ('next', ('agents', 'state_value')) OR ('agents', 'state_value') at the end of the trajectory: Value estimate of the state *at the end of the step*.
                  This should include the value estimate for the state after the last step.

        Returns:
            TensorDictBase: The input tensordict with 'advantage' and 'value_target' keys added.
        """
        # --- DEBUG PRINT: Inspect input tensordict keys ---
        # print("\n--- Inside CustomGAE forward ---") # Commented out for cleaner output during training
        # print("Input tensordict keys:", tensordict.keys(include_nested=True)) # Commented out
        # print(f"Input tensordict shape: {tensordict.shape}") # Commented out
        # --- END DEBUG PRINT ---


        # Ensure data is on the correct device (matching the GAE module's device)
        # This is important if the GAE module is moved to GPU but data is on CPU.
        # The input tensordict should already be on the correct device from the training loop,
        # but adding .to(self.device) here again is harmless and adds robustness.
        tensordict = tensordict.to(self.device)
        # print(f"CustomGAE: Input tensordict moved to device: {tensordict.device}") # Commented out


        # Extract necessary tensors and ensure they are on the correct device
        # Try both nested and flattened keys for robustness, prioritizing nested from collector
        rewards = tensordict.get(('next', ('agents', 'reward')), tensordict.get(('agents', 'reward'), None))
        if rewards is not None: rewards = rewards.to(self.device)

        terminated = tensordict.get('terminated')
        if terminated is not None: terminated = terminated.to(self.device)

        truncated = tensordict.get('truncated')
        if truncated is not None: truncated = truncated.to(self.device)

        values = tensordict.get(('agents', 'state_value'))
        if values is not None: values = values.to(self.device)

        # Special handling for next_values: need the value of the state *after* the last step in the trajectory.
        # This is expected at ('next', ('agents', 'state_value')) at the last index,
        # or potentially at the root ('agents', 'state_value') if the tensordict is not sliced
        # and represents a single step + next state.
        # Based on collector output structure, it should be at ('next', ('agents', 'state_value'))
        # within the collected batch tensordict.
        next_values_full_trajectory = tensordict.get(('next', ('agents', 'state_value')), None)
        if next_values_full_trajectory is not None: next_values_full_trajectory = next_values_full_trajectory.to(self.device)


        # Check if required keys were found
        if rewards is None:
             raise KeyError("CustomGAE: Could not find rewards at ('next', ('agents', 'reward')) or ('agents', 'reward').")
        if terminated is None:
             raise KeyError("CustomGAE: Could not find 'terminated' flag.")
        if truncated is None:
             raise KeyError("CustomGAE: Could not find 'truncated' flag.")
        if values is None:
             raise KeyError("CustomGAE: Could not find values at ('agents', 'state_value').")
        if next_values_full_trajectory is None:
             raise KeyError("CustomGAE: Could not find next_values at ('next', ('agents', 'state_value')).")


        # Combine terminated and truncated flags into a single done flag for GAE calculation
        # The done flags are for the state at time t+1, corresponding to the transition from time t.
        # In the collected data (shape [num_envs, frames_per_batch]), the flags at index `t`
        # are for the state at time `t+1`.
        dones = terminated | truncated # Shape [num_envs, frames_per_batch]
        # dones is already on the correct device because terminated and truncated were moved


        # GAE calculation: Iterate backwards from T-1 to 0.
        # advantage_t = reward_{t+1} + gamma * (1 - done_{t+1}) * value_{t+1} - value_t

        # We need rewards_{1...T}, done_{1...T}, values_{0...T-1}, next_values_{1...T}.
        # From collected data (shape [num_envs, frames_per_batch]), we have:
        # rewards_{1...T} are at index 0...frames_per_batch-1 in `rewards`. Shape [num_envs, frames_per_batch, num_agents, 1]
        # done_{1...T} are at index 0...frames_per_batch-1 in `dones`. Shape [num_envs, frames_per_batch]
        # values_{0...T-1} are at index 0...frames_per_batch-1 in `values`. Shape [num_envs, frames_per_batch, num_agents, 1]
        # next_values_{1...T} need careful indexing.
        # For t = 0 to T-2, next_value_{t+1} is `values[:, t+1]`.
        # For t = T-1, next_value_T is `next_values_full_trajectory[:, -1]`.

        # Create a tensor for next_values that aligns with the rewards, dones, and values tensors.
        # It should have shape [num_envs, frames_per_batch, num_agents, 1].
        # For indices 0 to frames_per_batch-2, use values[:, 1:].
        # For the last index (frames_per_batch-1), use next_values_full_trajectory[:, -1:].

        # Need to handle the batch shape potentially being more than 2 dimensions (e.g., if num_envs > 1)
        # The time dimension is assumed to be the second last dimension in the collected batch.
        # Let batch_dims = tensordict.batch_dims # Number of batch dimensions, typically 2 ([num_envs, frames_per_batch])
        # time_dim_idx = batch_dims - 1 # Index of the time dimension

        # Let's assume batch shape is [num_envs, frames_per_batch] for now.
        # Indexing for the time dimension is the second dimension (index 1).

        # Get values for time steps 1 to T-1
        values_t_plus_1_slice = values[:, 1:] # Shape [num_envs, frames_per_batch-1, num_agents, 1]

        # Get the last next value (for time step T)
        next_value_T = next_values_full_trajectory[:, -1:] # Shape [num_envs, 1, num_agents, 1]

        # Concatenate to get all next values from time step 1 to T
        next_values_aligned = torch.cat([values_t_plus_1_slice, next_value_T], dim=1)
        # Shape [num_envs, frames_per_batch, num_agents, 1] - This is values_{1...T}


        # Ensure done flags are broadcastable to agent dimension
        dones_expanded = dones.unsqueeze(-1).unsqueeze(-1).expand_as(rewards) # Shape [num_envs, frames_per_batch, num_agents, 1]


        # Calculate deltas: delta_t = reward_{t+1} + gamma * (1 - done_{t+1}) * value_{t+1} - value_t
        # Using our aligned tensors:
        # rewards: rewards_{1...T}
        # dones_expanded: done_{1...T} (expanded)
        # next_values_aligned: value_{1...T} (extracted from values and next_values_full_trajectory)
        # values: value_{0...T-1}

        # The delta calculation needs:
        # rewards[:, t] (reward_{t+1})
        # dones_expanded[:, t] (done_{t+1} expanded)
        # next_values_aligned[:, t] (value_{t+1})
        # values[:, t] (value_t)

        deltas = rewards + self.gamma * (1 - dones_expanded.float()) * next_values_aligned - values
        # Shape of deltas: [num_envs, frames_per_batch, num_agents, 1]


        # Calculate advantages using discounted sum of deltas (backward pass)
        advantages = torch.zeros_like(deltas) # Shape [num_envs, frames_per_batch, num_agents, 1]
        last_advantage = torch.zeros_like(advantages[:, 0]) # Shape [num_envs, num_agents, 1] # Represents advantage_{t+1}

        # Iterate backwards through time steps from T-1 down to 0
        for t in reversed(range(rewards.shape[1])):
             # Current delta is deltas[:, t]
             # The done flag at index t is for the state at time t+1.
             # This flag determines if advantage_{t+1} (last_advantage) should be zeroed out.
             # done_t_plus_1_expanded = dones_expanded[:, t] # Shape [num_envs, num_agents, 1]

             # GAE formula: advantages_t = delta_t + gamma * lmbda * (1 - done_{t+1}) * advantages_{t+1}
             # Using our indexing: advantages[:, t] = deltas[:, t] + gamma * lmbda * (1 - dones_expanded[:, t]) * last_advantage

             current_advantage = deltas[:, t] + self.gamma * self.lmbda * (1 - dones_expanded[:, t].float()) * last_advantage
             advantages[:, t] = current_advantage
             last_advantage = current_advantage # Update last_advantage for the next iteration


        # Value targets: value_target_t = advantage_t + value_t
        value_targets = advantages + values

        # Add advantages and value targets to the input tensordict
        tensordict.set('advantage', advantages)
        tensordict.set('value_target', value_targets)

        # print("CustomGAE: Added 'advantage' and 'value_target' to tensordict.") # Commented out for cleaner output during training
        # print("CustomGAE: Output tensordict keys:", tensordict.keys(include_nested=True)) # Commented out


        return tensordict
```

```python
#@title Pipeline Setup and Instantiation

import torch
import torch.nn as nn
import torch.optim as optim # Import optim
import torch_geometric.nn as gnn
from tensordict import TensorDict
from tensordict.nn import TensorDictModule, TensorDictSequential
from torchrl.modules import MLP, ProbabilisticActor
from torchrl.envs.utils import ExplorationType, set_exploration_type
from torchrl.collectors import SyncDataCollector # Import SyncDataCollector
from torchrl.data.replay_buffers import TensorDictReplayBuffer # Import TensorDictReplayBuffer
from torchrl.data.replay_buffers.storages import LazyTensorStorage # Import LazyTensorStorage
from torchrl.data.replay_buffers.samplers import SamplerWithoutReplacement # Import SamplerWithoutReplacement
from torchrl.objectives import PPOLoss # Import PPOLoss
from torch.utils.tensorboard import SummaryWriter # Import SummaryWriter
import os # Import os for path joining

# Ensure necessary classes and helper functions are available from previous cells
try:
    from __main__ import (
        FuelpriceenvfeatureGraph, # For data loading
        AnFuelpriceEnv,           # Environment
        PolicyBase,               # Base Policy Module
        ValueBase,                # Base ValueBase Module
        ForecastingBase,          # Base Forecasting Module
        CombinedPolicyForecastingBase, # Combined Module Base
        CustomGAE,                # Custom GAE Value Estimator
        multi_categorical_maker,  # Distribution maker
        ManualCollectionModule,   # Custom collector module wrapper
         _compute_advantages, # Helper function
         _optimize_step,      # Helper function
         _log_metrics,        # Helper function
         _save_models,        # Helper function
         print_tensordict_devices
    )
    print("All necessary classes and helper functions imported from global namespace for Pipeline Setup.")
except ImportError as e:
    print(f"Error importing necessary components for Pipeline Setup: {e}. Please ensure all previous cells defining these components have to be executed and were successful.")
    # Set essential components to None if import fails to prevent further errors
    FuelpriceenvfeatureGraph = None
    AnFuelpriceEnv = None
    PolicyBase = None
    ValueBase = None
    ForecastingBase = None
    CombinedPolicyForecastingBase = None
    CustomGAE = None
    multi_categorical_maker = None
    ManualCollectionModule = None
    _compute_advantages = None
    _optimize_step = None
    _log_metrics = None
    _save_models = None
    print_tensordict_devices = None
    print("Pipeline Setup cannot proceed due to missing components.")


# Ensure setup variables are defined from previous cells (same checks as in the training loop)
missing_components_setup = []
try:
    _ = hidden_rnn_dim
except NameError:
    hidden_rnn_dim = 64
    missing_components_setup.append("hidden_rnn_dim (using default)")

try:
    _ = input_x_dim # Should come from env.node_feature_dim or data_graph_loader.obs_dim
except NameError:
    input_x_dim = 14 # Default value including time
    missing_components_setup.append("input_x_dim (using default)")

try:
    _ = num_agents
except NameError:
    num_agents = 13
    missing_components_setup.append("num_agents (using default)")

try:
    _ = num_individual_actions_features
except NameError:
    num_individual_actions_features = 13
    missing_components_setup.append("num_individual_actions_features (using default)")

try:
    _ = num_action_categories
except NameError:
    num_action_categories = 3
    missing_components_setup.append("num_action_categories (using default)")

try:
     _ = forecast_horizon
except NameError:
     forecast_horizon = 30
     missing_components_setup.append("forecast_horizon (using default)")

try:
    _ = device
except NameError:
    device = 'cpu'
    missing_components_setup.append("device (using default)")

try:
    _ = env # Need env for action_spec and observation_spec
except NameError:
    env = None
    missing_components_setup.append("env")

try:
    _ = lr # Needed for optimizer
except NameError:
    lr = 3e-4 # Default value
    missing_components_setup.append("lr (using default)")

try:
    _ = clip_epsilon # Needed for PPOLoss
except NameError:
    clip_epsilon = 0.2 # Default value
    missing_components_setup.append("clip_epsilon (using default)")

try:
    _ = gamma # Needed for GAE
except NameError:
    gamma = 0.99 # Default value
    missing_components_setup.append("gamma (using default)")

try:
    _ = lmbda # Needed for GAE
except NameError:
    lmbda = 0.95 # Default value
    missing_components_setup.append("lmbda (using default)")

try:
    _ = frames_per_batch # Needed for buffer size and collector
except NameError:
    frames_per_batch = 100 # Default value
    missing_components_setup.append("frames_per_batch (using default)")

try:
    _ = total_frames # Needed for collector
except NameError:
    total_frames = 1000000 # Default value
    missing_components_setup.append("total_frames (using default)")

try:
    _ = minibatch_size # Needed for buffer sampling
except NameError:
    minibatch_size = 32 # Default value
    missing_components_setup.append("minibatch_size (using default)")

try:
    _ = max_grad_norm # Needed for optimizer
except NameError:
    max_grad_norm = 1.0 # Default value
    missing_components_setup.append("max_grad_norm (using default)")

try:
    _ = num_epochs # Needed for training loop
except NameError:
    num_epochs = 4 # Default value
    missing_components_setup.append("num_epochs (using default)")


if missing_components_setup:
    print(f"Warning: Some essential variables for module setup were not defined: {', '.join(missing_components_setup)}. Attempting to proceed with defaults where possible.")


# --- Re-instantiate components if necessary (due to previous errors or skipped cells) ---
# This block attempts to re-create necessary objects if they are None,
# relying on the classes and variables defined in previous cells.

# Attempt to re-instantiate environment if it's None
# Check if env is None AND if AnFuelpriceEnv class is available AND fuel_env_graph instance exists
if ('env' not in locals() or env is None) and AnFuelpriceEnv is not None and 'data_graph_loader_instance' in locals() and data_graph_loader_instance is not None:
    print("\nEnvironment not found or is None. Attempting re-instantiation using existing data_graph_loader_instance...")
    try:
        # Check if data_graph_loader_instance has combined_data and obs_dim set from a previous successful run
        if hasattr(data_graph_loader_instance, 'combined_data') and data_graph_loader_instance.combined_data is not None and \
           hasattr(data_graph_loader_instance, 'obs_dim') and data_graph_loader_instance.obs_dim is not None:

             data_tensor_for_env = data_graph_loader_instance.combined_data # Use the processed data tensor
             # input_x_dim is needed for module setup, get it from data loader
             input_x_dim = data_graph_loader_instance.obs_dim # Use obs_dim which includes time
             print(f"Inferred input_x_dim from data_graph_loader_instance: {input_x_dim}")

             # Ensure other env related variables are defined (use defaults if needed)
             episode_length = locals().get('episode_length', 10) # Use a shorter episode length for testing
             allow_repeat_data = locals().get('allow_repeat_data', True)
             num_envs = locals().get('num_envs', 3) # Ensure num_envs is defined
             seed = locals().get('seed', 42) # Ensure seed is defined
             device = locals().get('device', "cpu") # Ensure device is defined
             num_agents = locals().get('num_agents', 13) # Ensure num_agents is defined
             hidden_rnn_dim = locals().get('hidden_rnn_dim', 64) # Ensure hidden_rnn_dim is defined


             # Explicitly instantiate the environment here
             env = AnFuelpriceEnv(data_tensor=data_tensor_for_env, num_envs=num_envs, seed=seed, device=device, episode_length=episode_length, num_agents=num_agents, allow_repeat_data=allow_repeat_data)
             print("Environment re-instantiated successfully using existing data_graph_loader_instance data.")

        else:
             print("Existing data_graph_loader_instance instance does not have combined_data or obs_dim set. Cannot re-instantiate environment using it.")
             data_tensor_for_env = None
             env = None # Ensure env is None if data is missing
             # input_x_dim remains its default or value from previous cell

    except NameError as e:
        print(f"Required variables or data_graph_loader_instance not defined during environment re-instantiation attempt: {e}. Cannot re-instantiate environment.")
        data_tensor_for_env = None
        env = None
        # input_x_dim remains its default or value from previous cell
    except Exception as e:
        env = None
        print(f"An unexpected error occurred during environment re-instantiation attempt: {e}")
        # Print detailed traceback for debugging
        import traceback
        traceback.print_exc()
        # input_x_dim remains its default or value from previous cell

elif 'env' in locals() and env is not None:
     print("\nEnvironment found and is available.")
     # Ensure input_x_dim is set from the existing env if possible
     # The env's node_feature_dim should now match data_graph_loader.obs_dim
     if hasattr(env, 'obs_dim'):
          # env.obs_dim is the observation dimension including time
          input_x_dim = env.obs_dim
          print(f"Confirmed input_x_dim from existing env (obs_dim): {input_x_dim}")

     elif 'data_graph_loader_instance' in locals() and hasattr(data_graph_loader_instance, 'obs_dim') and data_graph_loader_instance.obs_dim is not None:
          # Fallback to data_graph_loader_instance if env.obs_dim is not available
          input_x_dim = data_graph_loader_instance.obs_dim
          print(f"Confirmed input_x_dim from existing data_graph_loader_instance (obs_dim): {input_x_dim}")
     else:
          # input_x_dim remains its default or value from previous cell
          print(f"Warning: Existing env or data_graph_loader_instance.obs_dim not available. Using default input_x_dim: {input_x_dim}")
else:
     print("\nEnvironment not found or is None, and cannot attempt re-instantiation.")
     env = None # Ensure env is None
     # input_x_dim remains its default or value from previous cell


# Instantiate Base Modules and Wrapped Modules
# Only proceed if base classes and essential variables are available AND env is available
if (PolicyBase is not None and ValueBase is not None and ForecastingBase is not None and
    CombinedPolicyForecastingBase is not None and ManualCollectionModule is not None and
    multi_categorical_maker is not None and env is not None):

    print("\nInstantiating Base Modules and Wrapped Modules...")
    try:
        # Instantiate base modules with correct input_x_dim
        # Ensure input_x_dim is correctly determined before this block
        policy_base = PolicyBase(input_x_dim=input_x_dim, num_agents=num_agents, num_individual_actions_features=num_individual_actions_features, num_action_categories=num_action_categories, hidden_rnn_dim=hidden_rnn_dim).to(device)
        value_base = ValueBase(input_x_dim=input_x_dim, num_agents=num_agents, hidden_rnn_dim=hidden_rnn_dim).to(device)
        forecasting_base = ForecastingBase(input_x_dim=input_x_dim, num_agents=num_agents, forecast_horizon=forecast_horizon, hidden_rnn_dim=hidden_rnn_dim).to(device)
        print("Base Modules instantiated.")

        # Instantiate combined module base
        combined_policy_forecasting_base = CombinedPolicyForecastingBase(
            policy_module_base=policy_base,
            forecasting_module_base=forecasting_base,
            hidden_rnn_dim=hidden_rnn_dim,
            num_agents=num_agents,
            device=device # Pass device to the combined module base
        ).to(device) # Move to device AFTER instantiation
        print("CombinedPolicyForecastingBase instantiated and moved to device.")


        # --- Instantiate TensorDictModule Wrappers ---
        # Wrap the combined policy/forecasting module base
        # Define input and output keys explicitly for TensorDictModule

        # Expected input keys to the combined_module_wrapper from the collector:
        # State data: ('agents', 'data', 'x'), ('agents', 'data', 'edge_index'), ('agents', 'data', 'batch'), ('agents', 'data', 'time')
        # RNN states from previous step: ('agents', 'rnn_hidden_state'), ('agents', 'rnn_hidden_state_forecast')

        combined_module_in_keys = [
            (('agents', 'data', 'x'), 'x'),
            (('agents', 'data', 'edge_index'), 'edge_index'),
            (('agents', 'data', 'batch'), 'batch'),
            (('agents', 'data', 'time'), 'time'),
            (('agents', 'rnn_hidden_state'), 'prev_policy_rnn_hidden_state'),
            (('agents', 'rnn_hidden_state_forecast'), 'prev_forecasting_rnn_hidden_state')
        ]
        print(f"Defined combined_module_wrapper in_keys: {combined_module_in_keys}")

        # Output keys for combined_module_wrapper (policy logits, next policy RNN, forecast, next forecasting RNN)
        # PolicyBase outputs unflattened logits: [batch_size, num_agents, num_individual_actions_features, num_action_categories]
        # Define out_keys as list of (tensordict_key, module_output_index_or_name) tuples
        combined_module_out_keys = [
             (('agents', 'logits'), 0), # First output is logits
             (('agents', 'rnn_hidden_state'), 1), # Second output is next policy RNN state
             ('forecast', 2), # Third output is forecast
             (('agents', 'rnn_hidden_state_forecast'), 3) # Fourth output is next forecasting RNN state
        ]
        print(f"Defined combined_module_wrapper out_keys: {combined_module_out_keys}")


        print("\nInstantiating combined_module_wrapper...")
        # Instantiate TensorDictModule for the combined policy/forecasting base
        combined_module_wrapper = TensorDictModule(
            module=combined_policy_forecasting_base,
            in_keys=combined_module_in_keys,
            out_keys=combined_module_out_keys,
        ).to(device)
        print("combined_module_wrapper instantiated.")
        print(f"combined_module_wrapper.in_keys: {combined_module_wrapper.in_keys}")
        print(f"combined_module_wrapper.out_keys: {combined_module_wrapper.out_keys}")


        # --- Instantiate Critic TensorDictModule Wrapper ---
        # Wrap the value module base
        # Input keys for critic_module from the collector:
        # State data: ('agents', 'data', 'x'), ('agents', 'data', 'edge_index'), ('agents', 'data', 'batch'), ('agents', 'data', 'time')
        # RNN state from previous step: ('agents', 'rnn_hidden_state_value')

        critic_in_keys = [
             (('agents', 'data', 'x'), 'x'),
             (('agents', 'data', 'edge_index'), 'edge_index'),
             (('agents', 'data', 'batch'), 'batch'),
             (('agents', 'data', 'time'), 'time'),
             (('agents', 'rnn_hidden_state_value'), 'prev_rnn_hidden_state') # Keyword matches ValueBase forward
        ]
        print(f"Defined critic_module in_keys: {critic_in_keys}")


        # Output keys for critic_module (state value, next value RNN)
        # Define out_keys as list of (tensordict_key, module_output_index_or_name) tuples
        critic_out_keys = [
             (('agents', 'state_value'), 0), # First output is state value
             (('agents', 'rnn_hidden_state_value'), 1) # Second output is next value RNN state
        ]
        print(f"Defined critic_module out_keys: {critic_module_out_keys}")


        print("\nInstantiating critic_module...")
        # Instantiate TensorDictModule for the value base
        critic_module = TensorDictModule(
            module=value_base,
            in_keys=critic_in_keys,
            out_keys=critic_out_keys,
        ).to(device)
        print("critic_module instantiated.")
        print(f"critic_module.in_keys: {critic_module.in_keys}")
        print(f"critic_module.out_keys: {critic_module.out_keys}")


        # List of learnable parameters for the optimizer
        learnable_params = []
        if policy_base is not None: learnable_params.extend(list(policy_base.parameters()))
        if value_base is not None: learnable_params.extend(list(value_base.parameters()))
        if forecasting_base is not None: learnable_params.extend(list(forecasting_base.parameters()))

        if learnable_params:
            print("\nLearnable parameters collected.")
            # Instantiate the optimizer
            print("\nInstantiating Optimizer...")
            optimizer = optim.Adam(learnable_params, lr=lr)
            print("Optimizer instantiated.")
        else:
            print("\nNo learnable parameters found. Skipping optimizer instantiation.")
            optimizer = None


        # --- Instantiate Loss Module ---
        print("\nInstantiating PPO Loss Module...")
        if combined_module_wrapper is not None and critic_module is not None:
             # Instantiate PPOLoss without the deprecated 'value_key' in constructor
             loss_module = PPOLoss(
                 actor=combined_module_wrapper,
                 critic=critic_module,
                 clip_epsilon=clip_epsilon,
                 entropy_bonus=0.01,
                 loss_critic_coef=0.5,
                 recurrent_state_keys=[('agents', 'rnn_hidden_state'),
                                       ('agents', 'rnn_hidden_state_forecast'),
                                       ('agents', 'rnn_hidden_state_value')],
                 distribution_keys=[('agents', 'logits')], # Explicitly tell PPOLoss where to find logits
                 # action_key is inferred from actor's output_spec or can be set here
                 # log_prob_key is inferred from action_key or can be set here
             ).to(device)

             # Use set_keys to specify the value_key after instantiation
             # The value_key should match where the critic_module outputs the value estimate.
             # Based on critic_module_out_keys, this is ('agents', 'state_value').
             loss_module.set_keys(value=('agents', 'state_value'))

             print("PPO Loss Module instantiated and value_key set.")
        else:
             loss_module = None
             print("Skipping PPO Loss Module instantiation due to missing combined_module_wrapper or critic_module.")


        # --- Instantiate Value Estimator (Custom GAE) ---
        print("\nInstantiating Custom GAE Value Estimator...")
        if CustomGAE is not None:
             # CustomGAE takes gamma and lmbda only in its constructor
             value_estimator = CustomGAE(
                 gamma=gamma,
                 lmbda=lmbda,
             ).to(device)
             print("Custom GAE Value Estimator instantiated.")
        else:
             value_estimator = None
             print("Skipping Custom GAE Value Estimator instantiation because CustomGAE class is missing.")


        # --- Instantiate Replay Buffer ---
        print("\nInstantiating Replay Buffer...")
        try:
             max_buffer_size = frames_per_batch * num_envs # Buffer size for one batch
             replay_buffer = TensorDictReplayBuffer(
                 storage=LazyTensorStorage(max_size=max_buffer_size, device=device), # Store on the specified device
                 sampler=SamplerWithoutReplacement(), # Sample the entire batch without replacement
                 batch_size=minibatch_size, # Minibatch size for sampling during optimization
             )
             print("Replay Buffer instantiated.")
        except Exception as e:
             print(f"An error occurred during Replay Buffer instantiation: {e}")
             # Print detailed traceback for debugging
             import traceback
             traceback.print_exc()
             replay_buffer = None


        # --- Instantiate Data Collector ---
        print("\nInstantiating Data Collector...")
        # Check if ManualCollectionModule class was successfully defined within this cell and other dependencies
        if (ManualCollectionModule is not None and combined_module_wrapper is not None and env is not None and
            _compute_advantages is not None and _optimize_step is not None and _log_metrics is not None and _save_models is not None and print_tensordict_devices is not None): # Check for necessary helper functions too
             try:
                 # Instantiate the ManualCollectionModule
                 # The ManualCollectionModule will now manually handle sampling using the maker
                 # It needs the combined_module_wrapper and the environment's action_spec.
                 policy_for_collector = ManualCollectionModule(
                     combined_module_wrapper=combined_module_wrapper,
                     action_spec=env.action_spec, # Pass the batched env action spec
                     device=device
                 ).to(device)
                 print(f"{ManualCollectionModule.__name__} instantiated for collector.")

                 # Instantiate SyncDataCollector
                 collector = SyncDataCollector(
                     env,
                     policy_for_collector, # Use the ManualCollectionModule as the policy
                     frames_per_batch=frames_per_batch,
                     total_frames=total_frames,
                     device=device, # Collector device
                     storing_device=device, # Device to store collected data
                     init_random_frames=0,
                     # The collector will handle observation keys based on the environment's spec
                     # and the policy's in_keys. No need for explicit observation_key.
                     # The collector automatically handles recurrent states if present in the policy's
                     # output tensordict under keys specified by the policy's state_keys property
                     # or inferred from the policy's output tensordict structure.
                 )
                 print("Data collector instantiated successfully.")
             except Exception as e:
                 print(f"An error occurred during SyncDataCollector instantiation: {e}")
                 # Print detailed traceback for debugging
                 import traceback
                 traceback.print_exc()
                 collector = None
        else:
             collector = None # Explicitly set collector to None if skipped
             print("Skipping Data Collector instantiation due to missing components:")
             if ManualCollectionModule is None:
                 print("- ManualCollectionModule class is missing or None.")
             if combined_module_wrapper is None:
                 print("- combined_module_wrapper is None.")
             if env is None:
                 print("- env is None.")
             if (_compute_advantages is None or _optimize_step is None or _log_metrics is None or
                 _save_models is None or print_tensordict_devices is None):
                 print("- Some refactored helper functions (_compute_advantages, _optimize_step, _log_metrics, _save_models, print_tensordict_devices) are missing or None.")


    except Exception as e:
        print(f"An unexpected error occurred during Module Setup and Instantiation: {e}")
        # Print detailed traceback for debugging
        import traceback
        traceback.print_exc()

else:
    policy_base = None
    value_base = None
    forecasting_base = None
    combined_module_wrapper = None
    critic_module = None
    learnable_params = []
    optimizer = None
    loss_module = None # Ensure loss_module is None
    value_estimator = None # Ensure value_estimator is None
    replay_buffer = None # Ensure replay_buffer is None
    collector = None # Ensure collector is None
    print("\nSkipping Module Setup and Instantiation because essential classes or environment are not available.")


# Debugging print statements to check the state of the base modules and pipeline components
print("\n--- Debugging: State of components after setup block ---")
print(f"policy_base is: {type(policy_base) if policy_base is not None else None}")
print(f"value_base is: {type(value_base) if value_base is not None else None}")
print(f"forecasting_base is: {type(forecasting_base) if forecasting_base is not None else None}")
print(f"combined_module_wrapper is: {type(combined_module_wrapper) if combined_module_wrapper is not None else None}")
print(f"critic_module is: {type(critic_module) if critic_module is not None else None}")
print(f"optimizer is: {type(optimizer) if optimizer is not None else None}")
print(f"loss_module is: {type(loss_module) if loss_module is not None else None}")
print(f"value_estimator is: {type(value_estimator) if value_estimator is not None else None}")
print(f"replay_buffer is: {type(replay_buffer) if replay_buffer is not None else None}")
print(f"ManualCollectionModule class is: {ManualCollectionModule}")
print(f"collector is: {type(collector) if collector is not None else None}")
print("--- End Debugging ---")
```

```python
#@title Training Loop

import time
import tqdm # Import tqdm for progress bar
from torch.utils.tensorboard import SummaryWriter # Already imported in pipeline setup

# Ensure necessary components are available from previous cells
try:
    from __main__ import (
        collector,
        value_estimator, # Import value_estimator
        replay_buffer,
        loss_module,
        optimizer,
        learnable_params, # List of learnable parameters
        policy_base, # For saving
        value_base,  # For saving
        forecasting_base, # For saving
        num_epochs,
        minibatch_size,
        max_grad_norm,
        log_interval,
        drive_path,
        num_agents,
        input_x_dim, # Needed for _optimize_step (though currently unused there)
        forecast_horizon, # Needed for _optimize_step (though currently unused there)
        device, # Needed for moving tensors to device
        _compute_advantages,
        _optimize_step,
        _log_metrics,
        _save_models,
        print_tensordict_devices,
        total_frames, # Make sure total_frames is available
        critic_module # Import critic_module
    )
    print("All necessary components for training loop imported from global namespace.")
except ImportError as e:
    print(f"Error importing necessary components for training loop: {e}. Please ensure all previous cells defining these components have been executed and were successful.")
    # Exit or set flags to prevent training if components are missing
    collector = None
    value_estimator = None # Ensure value_estimator is None if import fails
    critic_module = None # Ensure critic_module is None if import fails
    print("Training loop cannot proceed due to missing components.")


# Check if essential components are available before starting the loop
if collector is None or value_estimator is None or replay_buffer is None or loss_module is None or optimizer is None or policy_base is None or value_base is None or forecasting_base is None or critic_module is None: # Add critic_module to check
    print("\nSkipping training loop execution due to missing essential components.")
else:
    print("\nStarting training loop...")

    # TensorBoard logging setup
    log_dir = os.path.join(drive_path, "runs", f"ppo_run_{int(time.time())}")
    writer = SummaryWriter(log_dir)
    print(f"TensorBoard logs will be written to: {log_dir}")

    # Training loop variables
    collected_frames = 0
    start_time = time.time()
    episode_rewards = []
    episode_lengths = []

    # Use tqdm for a progress bar based on total_frames
    pbar = tqdm.tqdm(total=total_frames, desc="Training Progress")

    # Iterate through data collected by the collector
    # The collector will yield batches until total_frames is reached
    for i, data in enumerate(collector):

        # Update tqdm progress bar
        pbar.update(data.numel()) # numel() gives total number of elements (frames) in the tensordict

        # Check if collected data is a TensorDictBase
        if not isinstance(data, TensorDictBase):
             print(f"Warning: Collected data is not a TensorDictBase (type: {type(data)}). Skipping batch.")
             continue # Skip this batch if it's not a TensorDict

        # Ensure collected data is on the correct device (collector should handle this, but safe check)
        data = data.to(device)

        # Compute advantages and value targets
        # _compute_advantages expects a batch with shape [num_envs, frames_per_batch, ...]
        # The collector yields data in this shape.
        # print_tensordict_devices(data, "Collected Data before Advantage Comp") # Debug print

        # Pass critic_module to _compute_advantages
        data_with_advantages = _compute_advantages(data, critic_module, value_estimator, num_envs, num_agents, device)

        # Check if advantage computation was successful
        if data_with_advantages is None:
             print("Skipping optimization step due to failed advantage computation.")
             continue # Skip optimization for this batch

        # print_tensordict_devices(data_with_advantages, "Data after Advantage Comp") # Debug print


        # Add collected data to the replay buffer
        # The replay buffer expects data to be flattened across the batch dimensions
        # from [num_envs, frames_per_batch, ...] to [num_envs * frames_per_batch, ...]
        # Use data_with_advantages as it contains the computed advantages and value targets.
        # Ensure data_with_advantages is contiguous before adding to buffer
        data_to_buffer = data_with_advantages.contiguous()
        # Flatten the outer batch dimensions: [num_envs, frames_per_batch, ...] -> [num_envs * frames_envs * frames_per_batch, ...]
        data_to_buffer = data_to_buffer.reshape(-1) # Flatten the outer batch dimensions

        # Check if data_to_buffer is a TensorDictBase before adding
        if isinstance(data_to_buffer, TensorDictBase):
            replay_buffer.extend(data_to_buffer)
            # print(f"Added {data_to_buffer.shape[0]} frames to replay buffer. Current size: {len(replay_buffer)}") # Debug print
        else:
            print(f"Warning: Data to buffer is not a TensorDictBase (type: {type(data_to_buffer)}). Cannot add to buffer.")
            continue # Skip optimization if data cannot be added to buffer


        # Optimize network for num_epochs
        if len(replay_buffer) >= minibatch_size: # Only optimize if buffer has enough data for at least one minibatch
             # print(f"\nStarting optimization for batch {i}...") # Debug print
             for epoch in range(num_epochs):
                 # Sample minibatches from the replay buffer
                 # The sampler is SamplerWithoutReplacement, so each sample is unique per epoch
                 for minibatch in replay_buffer:
                     # Minibatch is already on the specified buffer device (which is the training device)
                     # print_tensordict_devices(minibatch, f"Minibatch (Epoch {epoch})") # Debug print

                     # Perform optimization step
                     # _optimize_step expects a minibatch with shape [minibatch_size, ...]
                     # It returns loss components
                     policy_loss, value_loss, entropy_bonus, forecasting_loss, total_loss = _optimize_step(
                         loss_module,
                         minibatch,
                         optimizer,
                         max_grad_norm,
                         learnable_params,
                         num_agents, # Pass num_agents
                         input_x_dim, # Pass input_x_dim
                         forecast_horizon, # Pass forecast_horizon
                         device # Pass device
                     )

                     # Log metrics for the minibatch
                     if total_loss is not None: # Only log if optimization step was successful
                         writer.add_scalar("Loss/Total", total_loss, collected_frames)
                         writer.add_scalar("Loss/Policy", policy_loss, collected_frames)
                         writer.add_scalar("Loss/Value", value_loss, collected_frames)
                         writer.add_scalar("Loss/Entropy", entropy_bonus, collected_frames)
                         writer.add_scalar("Loss/Forecasting", forecasting_loss, collected_frames) # Log forecasting loss


        # Update collected frames count after processing the batch
        collected_frames += data.numel()

        # Log training metrics periodically based on collected frames
        _log_metrics(i, collected_frames, total_loss, policy_loss, value_loss, entropy_bonus, forecasting_loss, data, log_interval, episode_rewards, episode_lengths, num_agents)


        # Check if total frames reached
        if collected_frames >= total_frames:
            break # Exit training loop

        # Clear the replay buffer after processing each batch (for on-policy PPO)
        replay_buffer.empty()
        # print("Replay buffer cleared.") # Debug print


    # --- Training Finished ---
    pbar.close()
    print("\nTraining finished.")

    # Close TensorBoard writer
    writer.close()
    print("TensorBoard writer closed.")

    # Save models
    _save_models(policy_base, value_base, forecasting_base, drive_path)

    # Optional: Visualize training progress (e.g., plot episode rewards/lengths)
    # This would require matplotlib or similar and is outside the core training loop.
    print("\nTraining complete. Models saved. You can now visualize training progress using TensorBoard.")
```
